```python
import sqlite3

# データをここに貼り付け
raw_data = """
@id,name,tourism,@lat,@lon
293705036,,viewpoint,35.2399643,139.0186005
304566283,スカイガーデン,viewpoint,35.4547727,139.6313256
391237968,,viewpoint,35.1974431,139.0258104
391237969,Lakeside Observation Building,viewpoint,35.1962174,139.0260842
391256329,,viewpoint,35.1955431,139.0237634
400905741,,viewpoint,35.3258899,139.5562973
400905794,,viewpoint,35.3252584,139.5565684
400905795,,viewpoint,35.3252122,139.5566547
404988090,,viewpoint,35.3240742,139.5555823
404988096,,viewpoint,35.3239767,139.5558158
428219221,,viewpoint,35.3234039,139.5564669
538950207,,viewpoint,35.6109905,139.5609100
597698437,下鶴間ふるさと館,museum,35.5008836,139.4642727
610687994,,viewpoint,35.4408584,139.6545296
610738967,,viewpoint,35.4405780,139.6546864
681242781,,viewpoint,35.3350725,139.5512498
681252192,,viewpoint,35.3354431,139.5511797
681252223,,viewpoint,35.3353204,139.5514451
681252232,,viewpoint,35.3351874,139.5512621
681268373,,viewpoint,35.3355717,139.5510291
681268382,,viewpoint,35.3358255,139.5510056
681268402,,viewpoint,35.3356789,139.5511410
681268421,,viewpoint,35.3354994,139.5510516
734220423,Kaze-no-oka Overlook,viewpoint,35.4972625,139.5232323
824017759,,viewpoint,35.6346130,139.2166404
824017762,,viewpoint,35.6296949,139.2207133
863002235,三菱みなとみらい技術館,museum,35.4557307,139.6300831
866264010,海上保安資料館 横浜館,museum,35.4542824,139.6442523
984966473,,viewpoint,35.4532499,139.6489720
1420789696,玉川大学教育博物館,museum,35.5693352,139.4727312
1420812637,そごう美術館,museum,35.4656520,139.6252040
1420813339,シルク博物館,museum,35.4472370,139.6452300
1420813409,ブリキのおもちゃ博物館,museum,35.4376390,139.6530630
1420813438,ミツトヨ博物館,museum,35.5965268,139.6204986
1420813521,三宮郷土博物館,museum,35.4013490,139.2844370
1420813557,横浜三渓園,museum,35.4170370,139.6587760
1420814483,伊勢原市立子ども科学館,museum,35.4039103,139.3140446
1420814728,光と緑の美術館,museum,35.5642870,139.3636570
1420814983,南足柄市郷土資料館,museum,35.3100480,139.0755580
1420815043,厚木市子ども科学館,museum,35.4409469,139.3664028
1420815356,厚木市郷土資料館(跡地),museum,35.4475007,139.3616093
1420815652,吉兆庵美術館,museum,35.3204156,139.5520620
1420816242,報徳博物館,museum,35.2489848,139.1535146
1420816450,大和市つる舞の里歴史資料館,museum,35.5136357,139.4542240
1420816548,大本山總持寺宝物殿,museum,35.5055386,139.6701145
1420816608,大磯町郷土資料館,museum,35.3054440,139.2908710
1420816676,大雄山最乗寺尚宝殿,museum,35.3021832,139.0762627
1420816707,女子美アートミュージアム,museum,35.5273934,139.3899608
1420816954,,museum,35.2510140,139.1534110
1420816964,小田原市尊徳記念館,museum,35.3025700,139.1456590
1420817098,尾崎咢堂記念館,museum,35.5942960,139.2463140
1420817112,山北町立丹沢森林館薬草園,museum,35.4142480,139.0676220
1420817119,山口蓬春記念館,museum,35.2663690,139.5761530
1420817137,山手資料館,museum,35.4383958,139.6526544
1420817157,岩崎博物館（ゲーテ座記念）,museum,35.4395140,139.6534679
1420817404,川崎市地名資料室,museum,35.5988110,139.6140520
1420817669,川崎市青少年科学館,museum,35.6082710,139.5615390
1420822202,平塚市美術館,museum,35.3380734,139.3478919
1420822426,徳富蘇峰記念館,museum,35.3026956,139.2627460
1420822506,愛川町郷土資料館,museum,35.5400200,139.2734670
1420822762,日本新聞博物館（ニュースパーク）,museum,35.4463516,139.6427009
1420822773,日本漫画博物館,museum,35.5866800,139.6445770
1420822813,日本郵船歴史博物館,museum,35.4503620,139.6394210
1420823012,木村美術館,museum,35.1588220,139.1095480
1420823351,棟方板画美術館,museum,35.3162530,139.5185200
1420823355,森のふれあい館,museum,35.1890840,139.0171220
1420823360,森の民話館,museum,35.4472500,139.3005880
1420823391,はまぎんこども宇宙科学館,museum,35.3770720,139.5949160
1420823424,横浜みなと博物館,museum,35.4527090,139.6326290
1420823425,横浜ユーラシア文化館,museum,35.4462299,139.6432801
1420824058,横浜市八聖殿郷土資料館,museum,35.4150750,139.6648130
1420825011,横浜市技能文化会館・技能文化展示室匠プラザ,museum,35.4418600,139.6350520
1420825110,横浜市歴史博物館,museum,35.5500860,139.5781790
1420825405,横浜市農村生活館・獅子ヶ谷横溝屋敷,museum,35.5218010,139.6427760
1420825565,横浜水道記念館,museum,35.4659261,139.5717610
1420825726,横浜都市発展記念館,museum,35.4461073,139.6434318
1420825914,横須賀市自然・人文博物館,museum,35.2745350,139.6737830
1420826079,江戸民具街道（おもしろ体験博物館）,museum,35.3241090,139.2192270
1420826157,津久井湖記念館,museum,35.5881190,139.2846140
1420826275,海外移住資料館,museum,35.4545787,139.6405733
1420827115,湯河原ゆかりの美術館,museum,35.1477930,139.0707820
1420827341,熊野郷土博物館,museum,35.5236540,139.6357080
1420828722,相模原市立博物館,museum,35.5592980,139.3931080
1420830282,神奈川県丹沢湖ビジターセンター,museum,35.4162072,139.0692704
1420830351,神奈川県水道記念館,museum,35.3715520,139.3796280
1420830414,神奈川県立地球市民かながわプラザ,museum,35.3680210,139.5532030
1420830415,神奈川県立埋蔵文化財センター,museum,35.4331710,139.6276030
1420830420,神奈川県立宮ケ瀬ビジターセンター,museum,35.5253480,139.2234600
1420830429,神奈川県立生命の星・地球博物館,museum,35.2393863,139.1215819
1420830439,神奈川県立金沢文庫,museum,35.3439291,139.6286449
1420830441,神奈川県立２１世紀の森,museum,35.3377810,139.0554660
1420831445,箱根武士の里美術館,museum,35.2677508,139.0059700
1420831578,箱根町立旧街道資料館,museum,35.2023825,139.0493339
1420831587,箱根町立郷土資料館,museum,35.2324689,139.1077112
1420831634,箱根関所資料館,museum,35.1935501,139.0262697
1420832021,羽太郷土資料館,museum,35.4373080,139.4763760
1420832375,若宮八幡宮郷土資料館,museum,35.5349410,139.7249900
1420832500,茅ヶ崎市文化資料館,museum,35.3223120,139.4044260
1420832580,茅ヶ崎市美術館,museum,35.3261730,139.4063440
1420832681,葉山しおさい博物館,museum,35.2641490,139.5779390
1420832785,藤沢市湘南台文化センターこども館,museum,35.3977760,139.4697110
1420833216,観音崎灯台資料展示室,museum,35.2561191,139.7450942
1420833224,観音崎自然博物館,museum,35.2517880,139.7423330
1420833427,辻堂海浜公園交通展示館,museum,35.3215109,139.4459330
1420833501,逗子市郷土資料館,museum,35.2874509,139.5765186
1420833815,鈴廣のかまぼこ博物館,museum,35.2458380,139.1310260
1420833921,鎌倉大谷記念美術館(休館),museum,35.3181930,139.5450240
1420834058,長谷寺宝物館,museum,35.3122938,139.5329577
1420834184,電気の史料館,museum,35.5338742,139.6728973
1420834189,電車とバスの博物館,museum,35.5878380,139.5924695
1420834306,馬の博物館,museum,35.4225483,139.6397486
1420834625,鶴岡八幡宮宝物殿,museum,35.3248660,139.5563110
1436512588,すすきの草原,viewpoint,35.2603653,139.0088361
1440505203,鐘,attraction,35.2656922,139.0173775
1447613578,,viewpoint,35.4627414,139.2950681
1447614193,物見峠,viewpoint,35.4645925,139.2963325
1447614778,むじな坂峠,viewpoint,35.4668344,139.2966890
1525926868,聖峰,viewpoint,35.4024000,139.2652253
1525932293,松田山　夜景スポット,viewpoint,35.3613400,139.1350631
1608807929,,attraction,35.6245248,139.5202083
1608808796,メリーゴーランドドッグ,attraction,35.6251501,139.5173363
1623963057,藤子・F・不二雄ミュージアム,museum,35.6100069,139.5735776
1633845141,のげやまこうえんてんぼうだい,viewpoint,35.4449876,139.6232658
1674457620,,viewpoint,35.3036628,139.6161260
1678499478,富士山とランドマークタワー,viewpoint,35.4418062,139.5620334
1707830135,展望ベンチ,viewpoint,35.3716782,139.2434820
1712857876,ペリー記念館,museum,35.2245922,139.7118988
1719284136,平塚テレビ中継局,viewpoint,35.3206767,139.3105445
1741692692,,viewpoint,35.3399950,139.2672316
1742559571,,viewpoint,35.3377789,139.2820587
1742559583,,viewpoint,35.3409438,139.2852238
1750408531,新幹線が見える丘,viewpoint,35.4093185,139.4078043
1777034847,ポケットパーク（旧西谷浄水場計量器室）,museum,35.4659735,139.5757294
1782561793,祇園山見晴台,viewpoint,35.3147638,139.5554223
1790785322,,viewpoint,35.3344893,139.5585603
1790785330,,viewpoint,35.3347528,139.5588412
1790785344,,viewpoint,35.3349743,139.5588944
1791271726,,viewpoint,35.3274989,139.5671119
1791272331,,viewpoint,35.3352275,139.5762402
1791638946,鶴岡文庫,museum,35.3272126,139.5564475
1826053372,パノラマ台,viewpoint,35.3109815,139.5674717
1826053374,貯水池,viewpoint,35.3110413,139.5685669
1826211444,景観百選(富士山),viewpoint,35.3128032,139.5682960
1838113117,新横浜ラーメン博物館,museum,35.5098496,139.6145269
1866713187,,viewpoint,35.3197845,139.3085784
1867297769,,viewpoint,35.4241396,139.3995544
1962033733,,viewpoint,35.3206722,139.3100534
1985922312,,viewpoint,35.5850682,139.1907195
2000929404,葉祥明美術館,museum,35.3352907,139.5491602
2008102149,富士岬平,viewpoint,35.4376803,138.9275733
2017858629,人間国宝美術館,museum,35.1531316,139.1107061
2038087727,見晴台,viewpoint,35.4345884,139.2452728
2046294047,,viewpoint,35.3388079,139.5499221
2068140284,富士見台,viewpoint,35.4355402,139.2317976
2090550553,鷹取山,viewpoint,35.6342086,139.1319304
2128808370,展望台,viewpoint,35.3216525,139.5890948
2192555516,,viewpoint,35.3064813,139.6126987
2204176349,遊行寺宝物館,museum,35.3479503,139.4877588
2207803875,秦野ビジターセンター,museum,35.4049111,139.1694152
2241101922,川崎マリエン,viewpoint,35.4964997,139.7632627
2241953404,,viewpoint,35.4246105,139.5955133
2253958411,原鉄道模型博物館,museum,35.4631490,139.6250863
2270165981,,viewpoint,35.2863967,139.0081480
2319204646,県庁本庁舎(King)屋上展望台,viewpoint,35.4479210,139.6429266
2320380959,しろくまの家,museum,35.4472715,139.6226601
2385383866,松見平休憩所,picnic_site;viewpoint,35.6092023,139.2688156
2540082950,,viewpoint,35.6054001,139.2786221
2604197602,薬師林道展望台,viewpoint,35.4438508,139.2837158
2604389803,天神平,viewpoint,35.4405958,139.2713045
2610525649,,viewpoint,35.4621465,139.2941734
2610883699,白山展望台,viewpoint,35.4716156,139.2985890
2639665169,,viewpoint,35.2501826,139.6278336
2646667707,仏果山山頂展望台,viewpoint,35.5207368,139.2581462
2749247357,寄神社,viewpoint,35.3971771,139.1392306
2763020275,バンデット,attraction,35.6255742,139.5181231
2763020278,ジャイアントスカイリバー,attraction,35.6243574,139.5206376
2763020279,レーザーアスレチック ファラオの罠,attraction,35.6260492,139.5181070
2763020283,アニマルレスキュー,attraction,35.6260473,139.5180987
2763020286,ゴーカート,attraction,35.6249376,139.5190297
2865743808,ななさわの丘,viewpoint,35.4491045,139.2998438
2865894776,ながめの丘,viewpoint,35.4526130,139.3016279
2865894777,もみじ展望台,viewpoint,35.4498893,139.3019152
2879272218,ヴェルニー記念館,museum,35.2847009,139.6559892
3028828138,Overlook,viewpoint,35.1982160,139.0264080
3048365027,宮川香山眞葛ミュージアム,museum,35.4680806,139.6297849
3076416568,皆城山展望台,viewpoint,35.3483069,139.5661552
3105693209,,viewpoint,35.5433231,139.2882621
3141111574,川崎・砂子の里資料館,museum,35.5320370,139.7018088
3193210404,畑宿一里塚,attraction,35.2112890,139.0621850
3202572868,,viewpoint,35.4362208,139.3168173
3202605616,,viewpoint,35.3766294,139.6346798
3202649002,,viewpoint,35.3773641,139.6278663
3209911101,高松山,viewpoint,35.4434105,139.3208281
3211950318,野鳥観察小屋B,viewpoint,35.3584058,139.6386264
3211950342,野鳥観察小屋A,viewpoint,35.3570570,139.6383896
3211950499,野鳥観察小屋C,viewpoint,35.3595378,139.6381733
3215545504,アクアライドII,attraction,35.3367753,139.6477549
3215545505,シーパラダイスタワー,attraction,35.3356158,139.6445732
3222748797,,viewpoint,35.4290277,139.6283151
3223134549,,viewpoint,35.4169734,139.6595567
3223134554,,viewpoint,35.4170182,139.6578779
3223134560,,viewpoint,35.4170588,139.6579345
3223134669,,viewpoint,35.4170929,139.6583519
3223134689,,viewpoint,35.4174034,139.6575132
3223134690,,viewpoint,35.4173842,139.6573703
3223134719,,viewpoint,35.4177648,139.6599614
3223134729,,viewpoint,35.4178796,139.6609800
3223134736,,viewpoint,35.4179213,139.6601763
3223134740,,viewpoint,35.4179362,139.6607481
3223134747,,viewpoint,35.4179630,139.6604821
3245631411,"KING, QUEEN & JACK TOWERS YOKOHAMA VIEW POINT",viewpoint,35.4473366,139.6433148
3245788081,"KING, QUEEN & JACK TOWERS YOKOHAMA VIEW POINT",viewpoint,35.4526876,139.6485330
3245788082,"KING, QUEEN & JACK TOWERS YOKOHAMA VIEW POINT",viewpoint,35.4509954,139.6449123
3293751041,,viewpoint,35.4245036,139.6612556
3362735297,厄除け門,attraction,35.5354023,139.7267155
3362873501,霊木 奇跡の銀杏,attraction,35.5345718,139.7298592
3362908048,川崎大師仲見世南から外観,attraction,35.5340330,139.7303320
3379278628,大山街道ふるさと館,museum,35.6033133,139.6127223
3385378400,一色海岸,viewpoint,35.2628585,139.5770578
3399389633,大和市郷土民家園,museum,35.4765696,139.4419249
3451517455,野鳥観察デッキ,viewpoint,35.4788270,139.4430946
3451517456,ひなた山展望広場,viewpoint,35.4790154,139.4446396
3485926669,展望デッキ,viewpoint,35.4766116,139.4452219
3518877275,池子遺跡群資料館,museum,35.3062722,139.5873207
3571877393,,viewpoint,35.2004099,139.0302648
3622839664,,viewpoint,35.2198580,139.6556591
3645334646,,viewpoint,35.2412801,139.7241978
3880226373,,viewpoint,35.1625910,139.6112298
3881822167,澤田痴陶人美術館,museum,35.1416057,139.6181143
3893023432,野鳥観察小屋Ｄ,viewpoint,35.3600087,139.6366992
3928709634,,viewpoint,35.1940783,139.6012427
3928709643,,viewpoint,35.1942639,139.6007566
3937261506,スラムダンク登場する場所,attraction,35.3066013,139.5020923
3942908354,披露山展望台,viewpoint,35.2959953,139.5643807
3942908355,大崎公園展望地,viewpoint,35.2934471,139.5592094
3942908356,逗子海岸展望地,viewpoint,35.2886832,139.5737792
3983444477,,viewpoint,35.1308718,139.6253216
3983444494,,viewpoint,35.1300959,139.6273198
4005625103,,viewpoint,35.2926461,139.5590657
4032187562,バードショー,attraction,35.5011675,139.5215799
4032259307,インドゾウ,attraction,35.4933984,139.5267685
4032282998,マレーバク,attraction,35.4938169,139.5255694
4032289538,ボルネオオランウータン,attraction,35.4932185,139.5262421
4032433036,スマトラトラ,attraction,35.4933855,139.5255886
4032444468,ウンピョウ,attraction,35.4929761,139.5258056
4032469109,ライオン,attraction,35.5010801,139.5202547
4064856333,,viewpoint,35.2798599,139.6387240
4071341618,高取山山頂展望台,viewpoint,35.5261751,139.2503001
4095566189,カワセミ,attraction,35.5046651,139.5361227
4095606565,ホタル,attraction,35.5023718,139.5336479
4095606566,ホタル,attraction,35.5029284,139.5333409
4116231327,水琴窟,attraction,35.4442539,139.6413327
4200311006,ウミウ展望台,viewpoint,35.1328072,139.6208339
4200311204,馬の背洞門,attraction,35.1312348,139.6183254
4225209213,Orbi,museum,35.4572991,139.6320915
4234939690,天王森泉館,viewpoint,35.3908775,139.4864528
4237276009,,viewpoint,35.4474157,139.2802648
4253713411,地球情報館,museum,35.3801876,139.6244823
4274597575,水車小屋,attraction,35.5684737,139.5010553
4293082150,見晴らし台,viewpoint,35.5584706,139.4954984
4300294489,汐入桟橋,viewpoint,35.2837836,139.6620021
4326967991,太閤風呂,attraction,35.2441858,139.0564811
4343867896,連合艦隊司令部 日吉地下壕 入口,viewpoint,35.5494186,139.6505609
4375125754,小田原城お濠の夜桜,attraction,35.2510879,139.1567499
4391368329,城山・かたくりの里,viewpoint,35.6017585,139.2967040
4392881019,,viewpoint,35.3807649,139.4181485
4394359103,,viewpoint,35.3732074,139.4177439
4394446154,チャレンジサイクル,attraction,35.5594772,139.4922471
4401247029,平塚新港フィシャリーナ,viewpoint,35.3144077,139.3643696
4401938976,観梅,viewpoint,35.3207509,139.5886375
4402110585,富士見の丘,viewpoint,35.5327959,139.6601682
4402130899,横浜港シンボルタワー,viewpoint,35.4389590,139.6896296
4403853920,渋田川沿いの芝桜,viewpoint,35.3965417,139.3377751
4403898067,鳶尾山観光展望台,viewpoint,35.4991328,139.3284839
4407519001,松風閣,viewpoint,35.4152033,139.6586830
4407526136,旧燈明寺三重塔,viewpoint,35.4163387,139.6592390
4427603602,,viewpoint,35.3026934,139.5243423
4428807392,,viewpoint,35.2995695,139.4807766
4428872089,,viewpoint,35.2990503,139.4747643
4434157542,箱根駅伝ミュージアム,museum,35.1893905,139.0248679
4435583889,黒崎の鼻,viewpoint,35.1829948,139.6151296
4438783410,文身歴史資料館,museum,35.4592595,139.6191536
4448911916,,viewpoint,35.2000870,139.0310618
4449902009,,viewpoint,35.2001984,139.0309594
4449955969,,viewpoint,35.2824737,139.6060080
4450424457,いこいの丘,viewpoint,35.2890404,139.6110488
4456086389,今熊神社,viewpoint,35.5651769,139.4012590
4456280289,菰釣山,viewpoint,35.4638156,138.9784308
4484888442,展望台,viewpoint,35.5863985,139.5196847
4493879290,,attraction,35.1971161,139.0385223
4496007891,展望広場,viewpoint,35.5313969,139.5526764
4518413633,,viewpoint,35.6086197,139.5576486
4529528146,ホタル,attraction,35.5676515,139.5002382
4544491796,見晴らし台,viewpoint,35.4957005,139.6149648
4624725379,粂野美術館,museum,35.5378438,139.5251328
4625371020,,viewpoint,35.3703749,139.5613725
4625371034,,viewpoint,35.3607535,139.5743085
4639728084,鶴見川流域センター,museum,35.5122532,139.5997238
4652538484,湯本茶屋一里塚,attraction,35.2281695,139.0968346
4652538485,葭原久保一里塚,attraction,35.1992638,139.0299524
4659818069,押切坂一里塚,attraction,35.2919147,139.2413927
4662881648,馬入一里塚跡碑,attraction,35.3313838,139.3605737
4662881651,小八幡一里塚,attraction,35.2749794,139.2023350
4662881652,山王原一里塚,attraction,35.2546342,139.1681266
4663300533,衣笠山公園展望台,viewpoint,35.2495751,139.6570318
4678585175,丸岳,viewpoint,35.2702602,138.9821936
4679536707,,viewpoint,35.3937856,139.5376877
4679536788,,viewpoint,35.3869326,139.5439465
4681425195,,viewpoint,35.2997609,139.4784859
4690618098,,viewpoint,35.3664668,139.5398949
4690618099,,viewpoint,35.3655471,139.5394912
4698962904,,viewpoint,35.3643756,139.5381024
4704828349,,viewpoint,35.3721350,139.5386394
4704828356,,viewpoint,35.3727035,139.5411823
4704828357,,viewpoint,35.3735061,139.5409396
4704828358,,viewpoint,35.3727330,139.5432984
4704828359,,viewpoint,35.3649942,139.5378894
4711743189,岡田美術館,museum,35.2383202,139.0463025
4716000959,,viewpoint,35.2867865,139.6960873
4728725293,,viewpoint,35.6071508,139.5847262
4729988826,見晴らし台,viewpoint,35.3000447,139.2012649
4730619818,神崎遺跡資料館,museum,35.4098148,139.4065852
4770852821,川喜多映画記念館,museum,35.3242675,139.5524688
4789032026,,viewpoint,35.6069536,139.5776017
4796173496,見晴らし山,viewpoint,35.4255938,139.6593918
4819521112,ポニー牧場,attraction,35.5603682,139.4949936
4826704675,長尾峠,viewpoint,35.2586117,138.9758349
4850938627,メリーゴーランド,attraction,35.4553847,139.6347044
4883311413,,viewpoint,35.2460085,139.0360163
4883312768,見張場,viewpoint,35.1924275,139.0270046
4883312769,,viewpoint,35.1926783,139.0268392
4890880740,,viewpoint,35.2441535,139.0202750
4890880741,,viewpoint,35.2443313,139.0203637
4894887031,王禅寺エコ暮らし環境館,museum,35.5930408,139.5247498
4895092861,OSANBASHI VIEWPOINT,viewpoint,35.4529820,139.6483673
4895092862,OSANBASHI VIEWPOINT,viewpoint,35.4509851,139.6466961
4908706770,,viewpoint,35.4370368,139.5666989
4924035837,象の鼻テラス,viewpoint,35.4496801,139.6431929
4946911464,アリじごく,attraction,35.5639129,139.4874323
5058590524,三浦桜祭り 臨時駐車場,viewpoint,35.1839620,139.6456742
5059644222,箱根ジオミュージアム,museum,35.2435748,139.0195210
5062217166,大田和 つつじの丘 駐車場,viewpoint,35.2320998,139.6425245
5080782082,電車とバスの博物館 B棟,museum,35.5880834,139.5933320
5141408824,,viewpoint,35.5179017,139.3290510
5156718778,,viewpoint,35.3301109,139.6000920
5182750902,彫刻の森美術館,museum,35.2442838,139.0520834
5261258385,,viewpoint,35.5404875,139.6301784
5261808626,,viewpoint,35.5487196,139.5241463
5265068007,ルーピングスターシップ,attraction,35.6245626,139.5197979
5265068015,わんデット,attraction,35.6248955,139.5182265
5265070632,プテラサイクル,attraction,35.6251426,139.5183580
5278956669,無線歴史展示室,museum,35.2236112,139.6738390
5287961022,古民家山十邸,attraction,35.5157194,139.3305873
5287961023,ふるさと資料館,viewpoint,35.5146999,139.3312742
5289210757,,viewpoint,35.5182439,139.6459367
5311092505,,viewpoint,35.3345999,139.5589244
5311094462,,viewpoint,35.3420933,139.5498283
5346571710,さざえ島,viewpoint,35.3005617,139.4855780
5385375669,D51広場,museum,35.4432972,139.3094867
5385825833,,viewpoint,35.2467479,139.1348298
5393130786,,viewpoint,35.3644312,139.5870669
5403438280,藤澤浮世絵館,museum,35.3406254,139.4477262
5541431521,,attraction,35.4529248,139.6347003
5545457988,,viewpoint,35.6125396,139.5584518
5565433825,,viewpoint,35.2778927,139.0530248
5603040065,駐車,attraction,35.4458940,139.6386230
5641247521,,attraction,35.2318935,139.0992338
5658824922,芦ノ湖遊覧船,attraction,35.2114698,139.0086433
5665244741,,attraction,35.3107555,139.5475385
5665302638,,attraction,35.3104239,139.5473095
5665302639,,attraction,35.3099339,139.5449550
5665302640,,attraction,35.3094882,139.5448504
5670448921,RCオアフ,attraction,35.5938169,139.5366087
5714116625,狩場JCT,attraction,35.4409347,139.5817848
5837313430,,viewpoint,35.2983927,139.4759067
5843806645,神奈川工科大学厚木市こども科学館,museum,35.4410395,139.3665927
5858769259,小原の里,museum,35.6228784,139.2011509
5965729039,,viewpoint,35.2993165,139.4780499
5966711491,見晴らし台,viewpoint,35.4262999,139.3226874
5971792817,,viewpoint,35.2642490,139.7281162
6036364460,海の見晴らし台,viewpoint,35.2542591,139.7427812
6067786385,,attraction,35.2488825,139.0182772
6067786585,,viewpoint,35.2491358,139.0184643
6068265386,,attraction,35.1514006,139.1427429
6070482408,,viewpoint,35.2579179,139.7340922
6103491571,塚山公園展望デッキ,viewpoint,35.2809817,139.6422520
6127216089,,attraction,35.2554813,139.1566245
6127216091,,attraction,35.3106213,139.4891430
6127218485,,attraction,35.2520742,139.1576423
6138382181,平賀敬美術館,museum,35.2308038,139.0988452
6138393562,,museum,35.2702605,139.0098665
6167529855,二代目ゴンドラ,attraction,35.2442247,139.0060096
6167529856,初代ゴンドラ,attraction,35.2442682,139.0059669
6199655998,,viewpoint,35.1968585,139.0259557
6204705282,,viewpoint,35.2434713,139.0199058
6292461085,,attraction,35.4439309,139.6480148
6320388246,かわさき宙と緑の科学館,attraction,35.6085010,139.5614498
6320388249,川崎市岡本太郎美術館,museum,35.6081367,139.5576721
6320388251,川崎市伝統工芸館,attraction,35.6088722,139.5571770
6320388252,はいさいフェスタ,attraction,35.5280778,139.6983193
6320388253,川崎市制記念多摩川花火大会,attraction,35.6049084,139.6307411
6320388255,瀋秀園,attraction,35.5317356,139.7323068
6320388256,自動車交通安全祈祷殿,attraction,35.5385684,139.7358655
6320388257,香林寺,attraction,35.6163080,139.5134498
6320388259,麻生不動院,attraction,35.5807045,139.5063984
6327216285,,viewpoint,35.2557432,139.0628136
6356069090,キリンビール横浜工場,attraction,35.4904335,139.6640692
6372225867,相模湖記念館,museum,35.6132664,139.1931333
6413944663,,viewpoint,35.3242998,139.5548077
6413944670,,viewpoint,35.3235571,139.5565765
6413944676,,viewpoint,35.3236602,139.5558997
6413945494,,viewpoint,35.3237069,139.5555248
6413945496,,viewpoint,35.3236780,139.5555778
6413945499,,viewpoint,35.3236232,139.5556770
6413945502,,viewpoint,35.3235549,139.5558889
6413945505,,viewpoint,35.3234695,139.5560291
6413945512,,viewpoint,35.3236851,139.5567331
6413945516,,viewpoint,35.3237789,139.5567574
6413945526,,viewpoint,35.3241371,139.5564705
6413945767,,viewpoint,35.3261311,139.5559255
6413945775,,viewpoint,35.3262066,139.5559868
6413945905,,viewpoint,35.3255569,139.5560861
6413945906,,viewpoint,35.3253557,139.5559637
6440912286,清水建設荏田ラグビー場,viewpoint,35.5554969,139.5633204
6447187242,あつぎ郷土博物館,museum,35.4953544,139.3498520
6490333885,,viewpoint,35.3116989,139.5326774
6516375969,,viewpoint,35.3071018,139.5647825
6516440400,,viewpoint,35.3076118,139.5649534
6550923551,かわさきエコ暮らし未来館,museum,35.5153529,139.7836219
6594152288,見晴し台,viewpoint,35.2525748,139.7414039
6700903686,曽我梅林,attraction,35.2995446,139.1930641
6700935386,原会場（原梅林）,attraction,35.2997472,139.1888360
6728721719,鶴巻温泉手湯「千の泉」,attraction,35.3810869,139.2769048
6815465511,展望デッキ,viewpoint,35.5753104,139.6594577
6840774584,ルーフテラス,viewpoint,35.3116506,139.4874322
6843903705,,viewpoint,35.2985254,139.4797953
6929058200,サクラの丘,viewpoint,35.2588610,139.1227651
6929109423,見晴らしの丘広場,viewpoint,35.2584885,139.1257375
6984040071,,viewpoint,35.3109209,139.5053430
6984040072,相模湾眺望,viewpoint,35.3095332,139.5077153
7021223804,うんこミュージアム YOKOHAMA,museum,35.4638344,139.6220899
7021699354,,viewpoint,35.3121587,139.5333448
7041347455,,viewpoint,35.1941123,139.6001293
7041347457,,viewpoint,35.1953245,139.5997551
7041347458,,viewpoint,35.1958176,139.6000502
7056021979,吉沢川紅葉,viewpoint,35.3283079,139.5792873
7107806008,,viewpoint,35.5811487,139.2815196
7140736861,箱根ドールハウス美術館,museum,35.2197822,139.0433061
7153785108,,viewpoint,35.2450510,139.0018279
7167954194,日本大学生物資源科学部 博物館,museum,35.3813045,139.4704029
7168171325,鎌倉オルゴール堂,museum,35.3127592,139.5344809
7265491186,,viewpoint,35.6296074,139.2207367
7265605621,,viewpoint,35.4176186,139.6610428
7275624324,やまゆり展望台,viewpoint,35.4439094,139.3034863
7277607325,,viewpoint,35.4929327,139.5238293
7278695940,,viewpoint,35.4976020,139.5221122
7281749302,土肥椙山巌窟(しとどの窟),attraction,35.1635716,139.0766040
7296854512,,viewpoint,35.4930477,139.5263823
7308316791,Pileated Gibbon,attraction,35.4935667,139.5260830
7308316792,インドライオン,attraction,35.4927617,139.5253395
7308316793,Lion-tailed Macaque,attraction,35.4925255,139.5256158
7372140404,富士見台,viewpoint,35.2789921,139.6402476
7503060812,,viewpoint,35.4572669,139.5380314
7554662686,服部牧場,viewpoint,35.5480941,139.2583396
7679398882,,viewpoint,35.3498327,139.5989091
7772325797,三浦半島　葉山町からの富士,viewpoint,35.2679210,139.5754181
7796157196,玉村豊男ライフアートミュージアム,museum,35.2035378,139.0308163
7821554686,,viewpoint,35.4925139,139.5227889
7847827522,,viewpoint,35.4932718,139.5268432
7847827531,,viewpoint,35.4933638,139.5268527
7847827540,,viewpoint,35.4930971,139.5269127
7847848493,,viewpoint,35.4926995,139.5260669
7847848494,,viewpoint,35.4924534,139.5255730
7847848497,,viewpoint,35.4925143,139.5258189
7847848504,,viewpoint,35.4927838,139.5260001
7847848505,,viewpoint,35.4925597,139.5259513
7847871181,,viewpoint,35.4935471,139.5260404
7847871490,,viewpoint,35.4933558,139.5257778
7847871512,,viewpoint,35.4932334,139.5261690
7847871513,,viewpoint,35.4935252,139.5256816
7847871522,,viewpoint,35.4930331,139.5258793
7847896720,,viewpoint,35.4929024,139.5239300
7847896729,,viewpoint,35.4928393,139.5251715
7847896732,,viewpoint,35.4926621,139.5254073
7847896740,,viewpoint,35.4924643,139.5232567
7847896745,,viewpoint,35.4927006,139.5241715
7847896750,,viewpoint,35.4927171,139.5252187
7847956099,,viewpoint,35.4925992,139.5225049
7847956100,,viewpoint,35.4941104,139.5238536
7847956104,,viewpoint,35.4944106,139.5240084
7847956106,,viewpoint,35.4945370,139.5240915
7847956108,,viewpoint,35.4932317,139.5224808
7847956110,,viewpoint,35.4933369,139.5232155
7847956115,,viewpoint,35.4937865,139.5236717
7847956123,,viewpoint,35.4931980,139.5225996
7847956124,,viewpoint,35.4942096,139.5238514
7847956126,,viewpoint,35.4933727,139.5232848
7847956129,,viewpoint,35.4926608,139.5221835
7847956133,,viewpoint,35.4936888,139.5240170
7847956137,,viewpoint,35.4943707,139.5239663
7847956146,,viewpoint,35.4931531,139.5227565
7847956147,,viewpoint,35.4945715,139.5240727
7847973697,,viewpoint,35.4953817,139.5228034
7847973699,,viewpoint,35.4952095,139.5233972
7847973701,,viewpoint,35.4953069,139.5227338
7847973705,,viewpoint,35.4945672,139.5236720
7847973710,,viewpoint,35.4946157,139.5237702
7847973711,,viewpoint,35.4953174,139.5233248
7848122957,,viewpoint,35.4970011,139.5243282
7848122983,,viewpoint,35.4970696,139.5241781
7848149422,,viewpoint,35.4957236,139.5235417
7848149425,,viewpoint,35.4959392,139.5227762
7848149432,,viewpoint,35.4957677,139.5234018
7848149433,,viewpoint,35.4956218,139.5237438
7848149441,,viewpoint,35.4954842,139.5238961
7848199902,,viewpoint,35.4988194,139.5221711
7848199914,,viewpoint,35.4979819,139.5221243
7848199935,,viewpoint,35.4989486,139.5216952
7848199942,,viewpoint,35.4990607,139.5222108
7848199946,,viewpoint,35.4989078,139.5217586
7848280125,,viewpoint,35.5004691,139.5213157
7848280129,,viewpoint,35.5011839,139.5211389
7848280131,,viewpoint,35.5007112,139.5212444
7848290644,,viewpoint,35.5011661,139.5198244
7848290648,,viewpoint,35.4996978,139.5211849
7848290650,,viewpoint,35.5003384,139.5202483
7848290654,,viewpoint,35.5013672,139.5205715
7848290665,,viewpoint,35.5005038,139.5204485
7848290672,,viewpoint,35.5012432,139.5198123
7848290674,,viewpoint,35.4997259,139.5207215
7848290676,,viewpoint,35.5007535,139.5206276
7848290679,,viewpoint,35.4997285,139.5212463
7848290680,,viewpoint,35.5002787,139.5204766
7848290683,,viewpoint,35.4997234,139.5209744
7848290684,,viewpoint,35.5015001,139.5206441
7848292386,,viewpoint,35.5002039,139.5207400
7848292387,,viewpoint,35.5014327,139.5209176
7852929481,,viewpoint,35.4976387,139.5222046
7852929482,,viewpoint,35.4977424,139.5222152
7852934392,,viewpoint,35.4982708,139.5221455
7852934393,,viewpoint,35.4983667,139.5220939
7852934394,,viewpoint,35.4984311,139.5220488
7853008587,,viewpoint,35.4931857,139.5223229
7853008588,,viewpoint,35.4931779,139.5223699
7853008592,,viewpoint,35.4931023,139.5220539
7853008593,,viewpoint,35.4928676,139.5221113
7853685768,,viewpoint,35.4925887,139.5226045
7853685773,,viewpoint,35.4926562,139.5230603
7853685775,,viewpoint,35.4924607,139.5235279
7853685776,,viewpoint,35.4924704,139.5233366
7853693774,,viewpoint,35.5011704,139.5200834
7853693775,,viewpoint,35.5009192,139.5200834
7855687973,,viewpoint,35.4930239,139.5265168
7855687974,,viewpoint,35.4930574,139.5265592
7855729388,,viewpoint,35.4930258,139.5263104
7855729404,,viewpoint,35.4937807,139.5256462
7856124133,,viewpoint,35.4946835,139.5229077
7856124135,,viewpoint,35.4948010,139.5227623
7856124136,,viewpoint,35.4949052,139.5227424
7856124137,,viewpoint,35.4950192,139.5226855
7856124138,,viewpoint,35.4952004,139.5226512
7856124143,,viewpoint,35.4952396,139.5226703
7859837711,,viewpoint,35.4929754,139.5237497
7859837712,,viewpoint,35.4925392,139.5236279
7859868801,,viewpoint,35.4955641,139.5238347
7875097935,,viewpoint,35.3366305,139.5477081
7875097938,,viewpoint,35.3366784,139.5476772
7875097940,,viewpoint,35.3368459,139.5478293
7875299469,,viewpoint,35.3383223,139.5484861
7875299480,,viewpoint,35.3383997,139.5486771
7875309200,,viewpoint,35.3384548,139.5486537
7875527540,北斎神奈川沖浪裏景観,viewpoint,35.4457102,139.6968235
7877313582,,viewpoint,35.3345601,139.5516225
7877320315,,viewpoint,35.3344824,139.5517387
7878462846,,viewpoint,35.4173799,139.6574529
7878610166,,viewpoint,35.4163305,139.6593342
7878790210,,viewpoint,35.4170638,139.6600153
7878790214,,viewpoint,35.4170357,139.6599777
7880366445,,viewpoint,35.4169277,139.6578748
7880386236,,viewpoint,35.4178170,139.6610332
7880386237,,viewpoint,35.4170616,139.6595345
7880386238,,viewpoint,35.4168612,139.6596847
7880386239,,viewpoint,35.4165757,139.6600536
7880386240,,viewpoint,35.4166681,139.6602888
7880386241,,viewpoint,35.4166327,139.6603308
7880594200,,viewpoint,35.4169608,139.6584120
7880594201,,viewpoint,35.4168994,139.6583690
7881095199,,viewpoint,35.4170975,139.6595233
7886120707,,viewpoint,35.3345001,139.5584959
7886120710,富士見台,viewpoint,35.3345270,139.5584479
7886120714,,viewpoint,35.3344408,139.5586377
7886120715,,viewpoint,35.3344145,139.5586877
7887360065,,viewpoint,35.4171283,139.6580969
7887360066,,viewpoint,35.4170880,139.6581468
7887360067,,viewpoint,35.4171169,139.6582797
7888537303,,viewpoint,35.4177722,139.6610621
7888673781,,viewpoint,35.4166115,139.6601228
7888673784,,viewpoint,35.4167153,139.6601488
7888681486,,viewpoint,35.4163061,139.6599903
7888681489,,viewpoint,35.4169463,139.6602949
7888681490,,viewpoint,35.4165053,139.6599322
7888807645,,viewpoint,35.4151978,139.6587461
7888809515,,viewpoint,35.4152775,139.6588034
7888809542,,viewpoint,35.4152208,139.6586580
7889184547,,viewpoint,35.4175020,139.6596445
7889184549,,viewpoint,35.4172141,139.6590196
7889184584,,viewpoint,35.4175407,139.6596796
7889201231,,viewpoint,35.4167470,139.6582077
7889201233,,viewpoint,35.4176277,139.6597471
7889201236,,viewpoint,35.4170378,139.6589859
7895433965,,viewpoint,35.3387355,139.5446770
7895462426,,viewpoint,35.3381797,139.5469133
7895462466,,viewpoint,35.3381055,139.5468247
7895484556,,viewpoint,35.3381894,139.5481968
7895498686,,viewpoint,35.3380209,139.5481436
7895498721,,viewpoint,35.3379732,139.5481900
7895797918,,viewpoint,35.3356838,139.5510742
7895797937,,viewpoint,35.3350711,139.5517497
7895798050,,viewpoint,35.3351741,139.5518312
7895798102,,viewpoint,35.3341980,139.5519320
7899197749,,viewpoint,35.3384988,139.5487876
7899197750,,viewpoint,35.3383557,139.5485865
7902003571,,viewpoint,35.3349669,139.5515135
7902003572,,viewpoint,35.3349304,139.5514210
7902003573,,viewpoint,35.3351367,139.5517181
7902003574,,viewpoint,35.3349691,139.5515701
7903947978,,viewpoint,35.3253833,139.5563208
7904752949,,viewpoint,35.3240407,139.5558999
7904752950,,viewpoint,35.3240119,139.5550376
7904752951,,viewpoint,35.3239321,139.5552336
7904752952,,viewpoint,35.3257366,139.5560072
7904752953,,viewpoint,35.3255473,139.5558412
7907590164,,viewpoint,35.3320822,139.5553545
7907593607,,viewpoint,35.3318431,139.5552151
7907593673,,viewpoint,35.3320330,139.5554236
7907593710,,viewpoint,35.3319194,139.5552227
7907593790,,viewpoint,35.3321313,139.5553063
7907593811,,viewpoint,35.3320033,139.5551376
7907766769,,viewpoint,35.3345071,139.5590430
7907766781,,viewpoint,35.3349811,139.5589592
7907840068,,viewpoint,35.3339805,139.5587373
7921943746,,viewpoint,35.3486507,139.4879794
7958450191,,viewpoint,35.3221581,139.5673523
7958450324,,viewpoint,35.3225164,139.5670919
7958450401,,viewpoint,35.3222006,139.5674753
7958450423,,viewpoint,35.3223457,139.5675586
7975496506,,viewpoint,35.3002515,139.4800736
7975561626,,viewpoint,35.2994260,139.4807377
7995517949,ふるさと上菅田　ささやま丘の上ミュージアム,museum,35.4925815,139.5652451
8000063721,,viewpoint,35.3224867,139.5674093
8000063722,,viewpoint,35.3223474,139.5673429
8083452782,赤星直忠博士文化財資料館,museum,35.2282361,139.6205041
8181782246,京急ミュージアム,museum,35.4623510,139.6262820
8181853767,,viewpoint,35.4515504,139.6358126
8181853768,,viewpoint,35.4508983,139.6358383
8181853769,,viewpoint,35.4509805,139.6353028
8181853770,,viewpoint,35.4514160,139.6362673
8326246817,,attraction,35.2828495,139.5687936
8338751155,東芝未来科学館,museum,35.5319890,139.6947127
8461561171,旧横濱鉄道歴史展示,museum,35.4498806,139.6316412
8508794797,,viewpoint,35.2370562,139.1283529
8509074782,,viewpoint,35.2353706,139.1280581
8566972614,真鶴町立遠藤貝類博物館,museum,35.1420563,139.1589623
8567584689,日産エンジンミュージアム,museum,35.4818216,139.6622914
8812072205,川崎浮世絵ギャラリー,museum,35.5319342,139.6985452
8952798489,,viewpoint,35.2998445,139.4745490
8960044369,,viewpoint,35.5835139,139.2895671
9014809562,原家旧住宅表門（陣屋門）／原家旧住宅稲荷社,attraction,35.5833341,139.6587380
9040064932,小杉御殿の御主殿跡,attraction,35.5837025,139.6541796
9042159681,磯部郷土民俗資料館,museum,35.5100188,139.3798764
9162046353,,viewpoint,35.3666016,139.4935986
9231714011,,viewpoint,35.5410056,139.7524645
9242911179,一国坂展望ポイント,viewpoint,35.2831781,139.6566666
9262697400,都筑中央公園　展望広場,viewpoint,35.5482734,139.5698027
9297842500,見晴し台,viewpoint,35.2680570,139.6438940
9310496390,化石島,attraction,35.6215962,139.5788610
9311816105,常磐富士見台,viewpoint,35.3240638,139.5270666
9344929713,,attraction,35.2846717,139.6559757
9385117318,スマイルサーキット（ラジコン）,attraction,35.5363351,139.2670877
9385117418,大野良ラジ場,viewpoint,35.4999406,139.3414026
9548487508,茅ヶ崎市開高健記念館,museum,35.3205502,139.4183954
9548487510,茅ヶ崎ゆかりの人物館,museum,35.3208738,139.4184082
9556266075,富士山と丹沢山地,viewpoint,35.4659584,139.5127181
9659278263,YUMESAKI GALLERY,museum,35.4608212,139.6261356
9697329191,さがみはらリニアブース,museum,35.5914421,139.3320735
9818271847,高山謝罪しろ!看板,attraction,35.3865635,139.2934043
10006253717,八景の棚,viewpoint,35.5219922,139.3775886
10131466717,ashiyu,attraction,35.2465076,139.0358266
10132628317,Torii,attraction,35.2238481,138.9990064
10132659717,平和の鳥居,attraction,35.2027411,139.0257425
10153737181,,viewpoint,35.2892896,138.9993381
10194492101,大観覧車,attraction,35.5548736,139.5783437
10216454044,横浜赤レンガ倉庫,attraction,35.4524046,139.6429182
10218913323,,viewpoint,35.2697697,138.9818445
10572334489,,viewpoint,35.2990943,139.4768238
10572344315,,viewpoint,35.2986975,139.4762997
10591674528,アトリエ展示室,museum,35.4538970,139.2999576
10770817072,,viewpoint,35.3688165,139.0497424
10772463840,,viewpoint,35.4582514,139.6374203
10783732905,,viewpoint,35.1979089,138.9941698
10802729122,,viewpoint,35.2432078,139.0186916
10802729123,,viewpoint,35.2436591,139.0187010
10823671605,,attraction,35.1961262,139.0280980
10823671705,,viewpoint,35.2048650,139.0210860
10823746905,,viewpoint,35.2435179,139.0186304
10931689469,展望台,viewpoint,35.3299096,139.4914289
11005779042,,viewpoint,35.3331742,139.4659769
11047504121,見晴台,viewpoint,35.4081139,139.4063582
11051153795,洋光台西公園見晴台,viewpoint,35.3813306,139.5919959
11062734991,世界の広場,attraction,35.4440847,139.6527207
11062761595,続日本100名城スタンプ125・小机城,attraction,35.5086091,139.5990728
11158073272,展望台,viewpoint,35.4972286,139.4031450
11171214500,,viewpoint,35.2762194,139.6730416
11263951466,,viewpoint,35.2588013,139.6088486
11263975578,,viewpoint,35.2593892,139.6064894
11266440294,,viewpoint,35.2559153,139.6056816
11266714777,,viewpoint,35.2235411,139.0272326
11266714802,,viewpoint,35.2233428,139.0242583
11276644013,月テラス,viewpoint,35.1453035,139.0737950
11276655797,川テラス,viewpoint,35.1452437,139.0726393
11379991304,半原水源地跡,attraction,35.5427251,139.2637629
11462328515,山田土筆細山美術館,museum,35.6151401,139.5097983
11500037820,Fuji View,viewpoint,35.5466445,139.5995427
11820466596,発句石,attraction,35.4955522,139.2896787
11896359569,第二展望台,viewpoint,35.1325410,139.6184244
11966015678,ハートの丘,viewpoint,35.3379047,139.2199882
11992126206,,viewpoint,35.3412380,139.2168185
12029220050,Ko-Kiridoshi,attraction,35.3281174,139.5930440
12029220220,Suda-ji,attraction,35.3260547,139.5914239
12029279409,O-Kiridoshi,attraction,35.3271445,139.5908137
12030075297,,attraction,35.3204230,139.5712492
12030075336,一条恵観山荘,museum,35.3203607,139.5713799
12055214426,,viewpoint,35.3550192,139.0799987
12076823714,,viewpoint,35.4855335,139.7453967
12174553041,,viewpoint,35.2768754,139.5821862
12310868794,,viewpoint,35.3226454,139.3240250
12336892201,唐沢線終点,viewpoint,35.4723102,139.2170787
12393236355,,viewpoint,35.1800214,139.0414982
12393241062,,viewpoint,35.1846707,139.0484348
12402915943,右卿記念館,museum,35.3300914,139.5579753
12443415526,,viewpoint,35.3612623,139.4920778
12452733530,,viewpoint,35.2990364,139.4745661
12452733531,Shōnan Monorail,viewpoint,35.3120944,139.4879223
12452733535,,viewpoint,35.2994324,139.4766414
12452733552,富士山,viewpoint,35.3292751,139.5417654
12519519136,愛の鐘,attraction,35.2185622,139.7062061
12530603870,,viewpoint,35.1563981,139.0886366
12533359372,,viewpoint,35.2208876,139.7101268
12533524303,,viewpoint,35.3676068,139.2232611
12533664535,,viewpoint,35.2222268,139.7045244
12533664557,,viewpoint,35.2191717,139.7015589
12570510611,,viewpoint,35.2793302,139.6415809
12603830501,Mt. Fuji Viewpoint,viewpoint,35.3031450,139.4815567
12603830601,Mt. Fuji Beach Viewpoint,viewpoint,35.3013188,139.4789235
12603830701,Waterfront,viewpoint,35.2991839,139.4747156
12620538997,かどやファームの扉,attraction,35.3827927,139.0507514
12630117456,,viewpoint,35.6217654,139.1838110
12637918114,しんかい2000,attraction,35.3103046,139.4791995
12645027665,山北町鉄道資料館,museum,35.3611425,139.0804336
12688858567,横濱ハイカラきもの館,museum,35.4473076,139.6450634
12691196123,森永エンゼルミュージアム MORIUM,museum,35.5172293,139.6737387
12698813583,,viewpoint,35.3051522,139.2901011
12722099899,自然観察センターしらかしの家,museum,35.4776201,139.4445972
12730272248,,viewpoint,35.3844190,139.2687872
12791577205,見晴茶屋,viewpoint,35.2085766,139.0571697
12794405060,,attraction,35.2535563,139.0497214
12807096042,,viewpoint,35.5455128,139.2357595
12810839750,雨岳文庫資料館,museum,35.4070282,139.2944456
12818393896,立野緑地,viewpoint,35.3590730,139.2266218
12839205601,,viewpoint,35.3370400,139.5730283
12839271001,,viewpoint,35.3371171,139.5737382
12862107201,展望広場,attraction,35.1769679,139.1155408
12961247576,見晴台,viewpoint,35.6054832,139.2434972
13004002001,箱根登山鉄道モハ1形107号,attraction,35.2459214,139.1307279
13006219155,早川橋梁,attraction,35.2341684,139.0871233
13012264311,Shinkansen bottom view at full speed,attraction,35.3907241,139.3723557
13014001330,cu―mo テラス,viewpoint,35.2464767,139.0359427
13114803301,蛇骨橋,viewpoint,35.2398445,139.0552387
13114811801,ミナカ小田原 展望足湯庭園,viewpoint,35.2547031,139.1553274
13135737407,さくらテラス,attraction,35.2846446,139.5947490
13176267312,大磯照ヶ崎のアオバト集団飛来地,attraction,35.3051432,139.3159731
13296000407,,viewpoint,35.4222395,139.3088237
13296000499,,viewpoint,35.4221057,139.3090836
13323554615,展望デッキ,viewpoint,35.2136210,139.7233449
13327750493,郷土資料館,museum,35.2442330,139.7150116
13398799601,Ajisai Bridge,attraction,35.2328479,139.1043127
13440969038,,viewpoint,35.5334019,139.7124205
13476447895,箱根写真美術館,museum,35.2499373,139.0460486
24573023,箱根駒ヶ岳ロープウェー,attraction,35.2184349,139.0170872
33039334,ドブ板通り,attraction,35.2819151,139.6657764
34742566,市立鎌倉国宝館,museum,35.3246687,139.5570209
34824401,メリーゴーラウンド,attraction,35.3376551,139.6431673
34824404,アクアミュージアム,attraction,35.3373631,139.6473263
36213480,横浜美術館,museum,35.4570856,139.6305994
46158374,神奈川県立近代美術館鎌倉別館,museum,35.3275664,139.5557557
47403128,神奈川近代文学館 本館,museum,35.4389275,139.6558072
47403131,山手111番館,museum,35.4387694,139.6547553
47403132,神奈川近代文学館 展示館,museum,35.4386463,139.6560668
47403138,大佛次郎記念館,museum,35.4393362,139.6553074
47403148,横浜市イギリス館,museum,35.4394095,139.6544356
47574836,横浜人形の家,museum,35.4431909,139.6521231
72970880,安藤百福発明記念館,museum,35.4554856,139.6388810
88383441,神奈川県立歴史博物館,museum,35.4491293,139.6362290
89951072,横浜大世界,theme_park,35.4425872,139.6480884
95908773,馬の博物館,museum,35.4229189,139.6396002
98891375,D51 408,attraction,35.6081382,139.5613503
130886674,箱根ガラスの森美術館,museum,35.2662614,139.0176593
130991780,氷川丸,attraction,35.4466672,139.6512412
131014913,すすき草原,attraction,35.2610719,139.0058461
138101264,日本丸,attraction,35.4537081,139.6323077
147428294,こどもの国,theme_park,35.5618511,139.4910567
147678780,,attraction,35.6241295,139.5207284
159045242,弘法山公園展望台,viewpoint,35.3718753,139.2430524
160552662,宮永岳彦記念美術館,museum,35.3817339,139.2765591
161338664,かながわアートホール,museum,35.4559100,139.5857560
163716723,麻布大学いのちの博物館,museum,35.5760388,139.3883776
164689197,よみうりランド,theme_park,35.6253010,139.5190652
167050302,小田原城歴史見聞館,museum,35.2506250,139.1557744
170957301,記念艦みかさ,museum,35.2851858,139.6743338
191557367,瑞心門,attraction,35.3004984,139.4801482
191660190,江島神社奉安殿,museum,35.3001648,139.4797608
194621307,北鎌倉古民家ミュージアム,museum,35.3363403,139.5467686
205604305,横浜外国人墓地資料館,museum,35.4395313,139.6523844
205727752,山手234番館,attraction,35.4376992,139.6521409
205727799,エリスマン邸,attraction,35.4375814,139.6512924
206375282,長井海の手公園 ソレイユの丘,theme_park,35.1950476,139.6118224
211888260,横浜市三殿台考古館,museum,35.4211132,139.6107644
232722954,小田原市郷土文化館,museum,35.2496759,139.1549582
236175972,,museum,35.6059911,139.5056904
239987847,旧石器ハテナ館,museum,35.5290148,139.3559611
244789179,神奈川県立近代美術館葉山,museum,35.2649927,139.5766483
250546705,,viewpoint,35.4418368,139.5621310
251760002,愛川町郷土資料館,museum,35.5399789,139.2553380
252651604,かわさき宙と緑の科学館,museum,35.6085480,139.5614207
253924492,海老名市温故館,museum,35.4540588,139.3973555
256127105,宝殿,museum,35.4396808,139.2748507
259292228,,viewpoint,35.5207332,139.2581417
261948827,真鶴町立中川一政美術館,museum,35.1444179,139.1518378
265574626,はだの歴史博物館,museum,35.3865047,139.1890593
278492513,東海道かわさき宿交流館,museum,35.5324742,139.7029569
291064691,城山ダム展望台,viewpoint,35.5857617,139.2849217
301753455,箱根町立郷土資料館,museum,35.2325065,139.1077105
312407308,平塚市博物館,museum,35.3369602,139.3486535
312734385,小田原市郷土文化館分館 松永記念館,museum,35.2490318,139.1414686
312734386,小田原文学館,museum,35.2447969,139.1536954
314206942,横浜・八景島シーパラダイス,theme_park,35.3362592,139.6456834
315103067,旧細菌検査室,museum,35.3573887,139.6363473
315107789,スプラッシュート,attraction,35.3365245,139.6424717
315107790,ドライブゾーン,attraction,35.3385745,139.6449481
315108198,うみファーム,attraction,35.3355764,139.6466672
315108199,シーボート,attraction,35.3364418,139.6463730
315292793,バイキング,attraction,35.3380784,139.6438761
315292801,オクトパス,attraction,35.3361705,139.6423892
315292805,ドランケン・バレル,attraction,35.3381647,139.6440837
315292808,サーフコースター　リヴァイアサン,attraction,35.3344743,139.6435422
315292809,ふれあいラグーン,attraction,35.3346037,139.6466847
315292813,ピーターパン,attraction,35.3380403,139.6436498
315292815,レッドバロン,attraction,35.3379203,139.6437748
319352562,オーシャン・スインガー,attraction,35.4548378,139.6339967
319352563,ディスク・オー,attraction,35.4553436,139.6336535
319352564,ギャラクシー,attraction,35.4556878,139.6343473
319352565,よこはまコスモワールド,theme_park,35.4556562,139.6355492
319352566,,attraction,35.4550305,139.6342740
319352568,,attraction,35.4552104,139.6369332
319352569,ファミリー・バナナ・コースター,attraction,35.4555391,139.6340169
319352571,スーパープラネット,attraction,35.4558301,139.6346817
327550627,箱根・芦ノ湖成川美術館,museum,35.2000610,139.0311946
351188081,山手68番館 (公園管理センター),attraction,35.4346179,139.6490698
351188083,横浜山手テニス発祥記念館,museum,35.4349142,139.6483242
352412031,やなぎテラス,viewpoint,35.1655248,139.6329923
363854124,コスモクロック21,attraction,35.4553703,139.6367666
366340575,民俗資料館 (旧三橋家),museum,35.3619815,139.4112394
366340576,民俗資料館 (旧和田家),museum,35.3625067,139.4123852
369306656,松前記念館,museum,35.3659045,139.2756718
370418216,,viewpoint,35.2841547,139.6297171
374688610,氷取沢ファミリー牧場,attraction,35.3592003,139.6018014
380067148,清川村宮ヶ瀬湖水の郷交流館,museum,35.5265456,139.2256658
381661095,ベーリックホール,attraction,35.4376450,139.6507137
388180518,ポーラ美術館,museum,35.2567097,139.0212216
390566734,一の鳥居,attraction,35.3132425,139.5485211
395998658,白秋記念館,museum,35.1364234,139.6219170
404722400,外交官の家,attraction,35.4366283,139.6432595
420038034,眺望テラス,viewpoint,35.1637211,139.6307592
427577141,大温室,museum,35.5656264,139.4875306
427577142,小温室,museum,35.5659804,139.4874471
439473845,,attraction,35.2594895,139.1264280
440428908,鏑木清方記念美術館,museum,35.3229806,139.5523324
445258887,横浜中華街,attraction,35.4428847,139.6451998
451336111,ブラフ18番館,attraction,35.4367830,139.6439217
455341527,,viewpoint,35.6061895,139.5586100
460141218,小町通り,attraction,35.3207885,139.5519476
467582787,小谷戸の里,museum,35.3866445,139.5487385
483423519,旧奥津邸,attraction,35.5188177,139.5177623
514503755,茅ヶ崎一里塚,attraction,35.3333696,139.4071568
527376225,日吉の森庭園美術館,museum,35.5594726,139.6347502
539541663,さがみ湖MORI MORI,theme_park,35.6054026,139.2076487
543027961,箱根神社宝物殿,museum,35.2040883,139.0248084
544683684,ひゅーどろ,attraction,35.6250434,139.5185101
550393197,,attraction,35.6248790,139.5181295
558462846,D51広場,museum,35.4432399,139.3095623
566223590,酒匂川ふれあい館,museum,35.3340413,139.1335370
568389897,鎌倉歴史文化交流館,museum,35.3224054,139.5475217
577520772,二ヶ領せせらぎ館,museum,35.6205280,139.5751123
578882071,川崎市岡本太郎美術館,museum,35.6078900,139.5574895
593859422,川崎市平和館,museum,35.5687866,139.6589535
600959322,海老名市立歴史資料収蔵館,museum,35.4494775,139.3763500
655201036,D51,museum,35.4432065,139.3096294
689595633,古民家山十邸,museum,35.5157591,139.3307694
702538918,BAM,museum,35.3226621,139.5545516
704524199,鎌倉文学館,museum,35.3152948,139.5389957
718076346,玉縄歴史館,museum,35.3512416,139.5194381
718511002,小町通り,attraction,35.3241952,139.5537031
723905273,横浜アンパンマンこどもミュージアム＆モール,theme_park,35.4632667,139.6289419
727045804,大涌谷,attraction,35.2467581,139.0250167
729414234,竹の庭,attraction,35.3197204,139.5687530
742439445,蒸気機関車 D52-235,attraction,35.5668203,139.3936404
753629108,海と夕日の広場,attraction,35.1946211,139.6106636
772690985,箱根関所,museum,35.1922482,139.0263584
775467804,ウォータースライダー,attraction,35.3220234,139.4513252
775467805,ウォータースライダー,attraction,35.3220261,139.4513258
775467817,,attraction,35.3217858,139.4455224
779480823,コアラ,attraction,35.3519883,139.5977702
779535086,シロテテナガザル,attraction,35.3501518,139.5964886
779535087,シロテテナガザル,attraction,35.3501010,139.5964026
799731426,,attraction,35.6239260,139.5206149
799731427,,attraction,35.6239437,139.5207261
819515883,,museum,35.5642859,139.3637147
822440721,ヒガシクロサイ,attraction,35.3500726,139.5973934
822440723,アミメキリン,attraction,35.3503417,139.5972644
822440724,アラビアオリックス,attraction,35.3505604,139.5971218
822440725,オカピ,attraction,35.3508559,139.5969496
822444503,アオバネワライカワセミ,attraction,35.3516210,139.5974503
822444504,オオカンガルー,attraction,35.3513241,139.5978497
822444508,オオツノヒツジ,attraction,35.3485926,139.5972532
822446078,スバールバルライチョウ,attraction,35.3488100,139.5971000
822449862,カピバラ,attraction,35.3480327,139.5967279
822449863,イノシシ,attraction,35.3479444,139.5965809
822449864,ニワトリ,attraction,35.3479257,139.5965233
822449865,ミニブタ,attraction,35.3478841,139.5964712
822449866,ポニー,attraction,35.3478520,139.5963345
822449867,ヒツジ,attraction,35.3478268,139.5961312
822449868,ヤギ,attraction,35.3477604,139.5961775
822453476,マーラ,attraction,35.3482271,139.5951006
822453479,ベアードバク,attraction,35.3484057,139.5951056
822453480,インドゾウ,attraction,35.3486670,139.5953554
822453482,インドサイ,attraction,35.3492269,139.5957250
822453483,インドサイ,attraction,35.3490811,139.5955895
822455069,ニホンカモシカ,attraction,35.3495777,139.5959845
822455070,スーチョワンバーラル,attraction,35.3496755,139.5962056
822455071,ホンシュウジカ,attraction,35.3504631,139.5962200
822455072,タンチョウ,attraction,35.3506098,139.5958400
835026590,鈴廣かまぼこ博物館,museum,35.2454972,139.1288851
835257713,石仏群と歴史館,museum,35.2129925,139.0357001
840547397,ロマンスカーミュージアム,museum,35.4529335,139.3900167
879791929,,museum,35.6020719,139.6260678
888197265,,museum,35.4952439,139.3498125
893061189,星のミュージアム,museum,35.5959044,139.6129497
915458374,横浜開港資料館,museum,35.4474991,139.6438036
921126881,第一展望台,viewpoint,35.1338889,139.6136120
952587264,菜の花台園地展望台,viewpoint,35.4100318,139.2086182
954471595,水とかがやく未来館,museum,35.6045357,139.5439739
964654518,本間寄木美術館,museum,35.2368033,139.1168101
973483833,札掛森の家,museum,35.4614014,139.2131627
985029771,スハ42-2047,attraction,35.6086290,139.5606682
1054387277,箱根美術館,museum,35.2481895,139.0428118
1074398924,ワイルドビーチ座間,theme_park,35.4995205,139.4206888
1083644085,茅ヶ崎市博物館,museum,35.3633846,139.4124896
1105675923,澤田美喜記念館,museum,35.3101845,139.3138550
1120516740,八幡山の洋館,museum,35.3337940,139.3477850
1121040519,光触媒ミュージアム,museum,35.5947672,139.6209674
1154160042,小町通り,attraction,35.3231612,139.5531592
1170049109,宇宙科学探査交流棟,museum,35.5589777,139.3946824
1182684218,展望ひろば,viewpoint,35.5937274,139.3446063
1214718113,,attraction,35.6238581,139.5206472
1214718114,,attraction,35.6239866,139.5206075
1214718115,,attraction,35.6240456,139.5206587
1214718116,,attraction,35.6239639,139.5208462
1214718117,,attraction,35.6240833,139.5207913
1214718118,,attraction,35.6241419,139.5204841
1214718119,,attraction,35.6240829,139.5205255
1216999681,段テラス,viewpoint,35.1449966,139.0734696
1217012358,蛍テラス,viewpoint,35.1451497,139.0723739
1252233173,,museum,35.5381282,139.5928350
1252233174,,museum,35.5381036,139.5926478
1252743040,,museum,35.5380812,139.5928892
1290365056,太陽の公園,viewpoint,35.3244378,139.2319450
1323466115,門沢橋出土品整理室,museum,35.4098587,139.3760828
1371652318,奥津国道美術館,museum,35.3815317,139.2200303
1375182298,SHODAIテラス,viewpoint,35.4982376,139.6599031
1382214779,,attraction,35.2570733,139.6400513
1382887991,,viewpoint,35.5261751,139.2503062
1383355449,,viewpoint,35.5455236,139.2357488
1383681430,横浜市電保存館,museum,35.4203915,139.6244946
1452577016,,viewpoint,35.4227079,139.3091045
1468565524,展望デッキ,viewpoint,35.3707594,139.2211795
8200233,日本民家園,museum,35.6093466,139.5596266
10359304,鎌倉文華館 鶴岡ミュージアム,museum,35.3247228,139.5548281
11580482,北斎富嶽三十六景,viewpoint,35.4457102,139.6968235
19193052,横須賀美術館,museum,35.2596682,139.7382639

"""

def build_database():
    conn = sqlite3.connect('attraction.db')
    cursor = conn.cursor()

    cursor.execute('DROP TABLE IF EXISTS attractions')
    cursor.execute('''
    CREATE TABLE attractions (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        category TEXT,
        lat REAL NOT NULL,
        lon REAL NOT NULL
    )
    ''')

    lines = raw_data.strip().split('\n')
    insert_data = []
    
    for line in lines:
        parts = line.split(',')
        if len(parts) < 4: continue # データが足りない行は飛ばす
        
        try:
            # 後ろから順に取得することで、名称にカンマが含まれていても対応可能
            lon = float(parts[-1])      # 最後
            lat = float(parts[-2])      # 最後から2番目
            category = parts[-3].strip() # 最後から3番目
            
            # 名称は「ID」と「カテゴリ」の間にあるものすべてを結合する
            # これにより、名称内のカンマも正しく復元されます
            name = ",".join(parts[1:-3]).strip()
            osm_id = int(parts[0])
            
            # 名称が空の場合（,,となっている場合）はカテゴリ名を仮の名前として使うかスキップ
            if not name:
                # 名前がないスポットはスキップする場合
                continue
                
            insert_data.append((osm_id, name, category, lat, lon))
            
        except ValueError:
            # 数値変換できない行（ヘッダーなど）があれば無視する
            continue

    cursor.executemany('''
    INSERT INTO attractions (id, name, category, lat, lon)
    VALUES (?, ?, ?, ?, ?)
    ''', insert_data)

    conn.commit()
    conn.close()
    print(f"成功！ {len(insert_data)} 件のスポットを登録しました。")

if __name__ == "__main__":
    build_database()
```

```python
import sqlite3

# 静岡県・山梨県のデータをここに貼り付け
raw_data = """
@id,name,tourism,@lat,@lon
416891983,,viewpoint,34.9703339,138.0617100
417314623,栗ヶ岳,viewpoint,34.8418401,138.0686102
567816055,天竜湖畔の森山頂,viewpoint,34.9068566,137.8121664
714929634,ビンカ峠,viewpoint,35.0488674,138.0474867
735512257,音止めの滝,attraction,35.3123823,138.5892235
771373068,八丁池展望台,viewpoint,34.8441536,138.9621932
777523388,香貫山展望台,viewpoint,35.0918612,138.8756948
828448474,静岡市民ギャラリー,museum,34.9754988,138.3823664
839992072,井川展示館,museum,35.2115409,138.2225107
874165763,六合目,attraction,35.3391773,138.7379053
874172117,新七合目,attraction,35.3447667,138.7369878
880714965,元祖七合目,attraction,35.3485217,138.7359413
880719327,八合目,attraction,35.3517537,138.7353998
888297714,九合目,attraction,35.3545767,138.7328192
888306879,九合五勺,attraction,35.3567748,138.7311862
888352418,六合目,attraction,35.3533294,138.7498385
888357831,七合目,attraction,35.3543573,138.7448369
888363475,七合四勺,attraction,35.3529863,138.7428089
890510432,七合九勺,attraction,35.3545330,138.7383454
890518723,八合目,attraction,35.3561132,138.7367878
890553086,新六合目,attraction,35.3509426,138.7552709
1275806523,中部電力浜岡原子力館,museum,34.6289480,138.1414807
1370878571,資料館,museum,35.3567622,138.8624413
1400767918,円山花木園,viewpoint,34.9622069,138.3522204
1402646535,砂払い五合目「下り」,viewpoint,35.3679849,138.7676393
1420763518,岡部宿大旅籠柏屋,museum,34.9192941,138.2826715
1420974999,さくま郷土遺産保存館,museum,35.0811140,137.7953590
1420975160,ねむの木子ども美術館,museum,34.6356530,138.1249430
1420975619,ビュフェ美術館・ビュフェこども美術館,museum,35.1715320,138.8695820
1420975622,フォーレなかかわね茶茗舘,museum,35.0704476,138.1078135
1420975637,メルヘンの森美術館,museum,34.9034170,139.1174050
1420975763,三島市郷土資料館,museum,35.1229168,138.9101657
1420975787,三嶋大社宝物館,museum,35.1214025,138.9197079
1420975983,上原仏教美術館,museum,34.7281560,138.9257610
1420975985,上原近代美術館,museum,34.7285598,138.9255743
1420976366,伊豆市資料館,museum,34.9643960,138.9844130
1420976417,中勘助文学記念館（杓子庵）,museum,34.9851780,138.3175200
1420977369,伊豆オルゴール館,museum,34.8770870,139.1036610
1420977378,伊豆一碧湖美術館,museum,34.9286955,139.1084201
1420977549,佐野美術館,museum,35.1149319,138.9159324
1420977565,ジオリア（伊豆半島ジオパークミュージアム）,museum,34.9725880,138.9334768
1420977568,修禅寺宝物館,museum,34.9711985,138.9279390
1420978134,加藤学園考古学研究所,museum,35.1173382,138.8689150
1420978654,吉岡彌生記念館,museum,34.7063790,138.0479610
1420978915,唐人お吉記念館,museum,34.6764310,138.9430550
1420979960,城ケ崎文化資料館（おもしろ博物館）,museum,34.8767350,139.1100920
1420981105,奇石博物館,museum,35.2818460,138.6397430
1420981526,富士川町立地方歴史民俗資料館,museum,35.1475030,138.6192360
1420981596,富士市立博物館,museum,35.1854450,138.6800765
1420982944,島田市博物館,museum,34.8313360,138.1543820
1420982952,島田市博物館分館,museum,34.8321598,138.1565786
1420983287,川根本町資料館やまびこ,museum,35.1787805,138.1864150
1420988257,平野美術館,museum,34.7156649,137.7311006
1420988419,御前崎市清川泰次芸術館,museum,34.5975480,138.2215360
1420988496,御殿場スポーツカーガーデン,museum,35.2870100,138.9526290
1420988916,掛川市二の丸美術館,museum,34.7757329,138.0144620
1420989007,新居宿旅籠・紀伊国屋資料館,museum,34.6946925,137.5598500
1420989024,新居関所史料館,museum,34.6948427,137.5616958
1420989069,日本人形博物館・日本招き猫美術館,museum,34.9798410,138.3793550
1420989232,金原明善記念館,museum,34.7281315,137.7912784
1420989541,東海大学海洋科学博物館,museum,35.0191780,138.5182380
1420989543,東海大学自然史博物館,museum,35.0186430,138.5187260
1420990534,梅蔭禅寺次郎長遺物館,museum,35.0053800,138.4862480
1420990680,森町立歴史民俗資料館,museum,34.8391231,137.9255107
1420990994,池田20世紀美術館,museum,34.9217330,139.1058579
1420991052,沼津市戸田造船郷土資料博物館,museum,34.9767711,138.7663971
1420991053,沼津市明治史料館・江原素六記念館,museum,35.1218534,138.8583389
1420991054,沼津市歴史民俗資料館,museum,35.0713880,138.8732660
1420991066,沼津市若山牧水記念館,museum,35.0870893,138.8516629
1420992450,浜松市天文台,museum,34.6634788,137.7689202
1420992456,浜松市姫街道と銅鐸の歴史民俗資料館,museum,34.8104550,137.6520838
1420992468,浜松市楽器博物館,museum,34.7066154,137.7382244
1420992490,浜松市秋野不矩美術館,museum,34.8648770,137.8193000
1420992516,浜松市立内山真龍資料館,museum,34.8770080,137.8143740
1420992637,浜松市舞阪郷土資料館,museum,34.6794484,137.6134163
1420993481,清水港湾博物館（フェルケール博物館）,museum,35.0075710,138.4931800
1420994033,焼津市歴史民俗資料館,museum,34.8645280,138.3053660
1420994089,焼津漁業資料館,museum,34.8711430,138.3239993
1420994497,熱海市立伊豆山郷土資料館,museum,35.1158700,139.0825950
1420994507,熱海市立澤田政廣記念美術館,museum,35.0988680,139.0578480
1420994729,牧之原市榛原郷土資料館,museum,34.7428387,138.2275633
1420994740,牧之原市相良史料館,museum,34.6859120,138.1979070
1420995344,玉泉寺ハリス記念館,museum,34.6757240,138.9624110
1420995378,玉澤妙法華寺宝物殿,museum,35.1231140,138.9626540
1420998630,磐田市旧見付学校,museum,34.7275080,137.8569910
1420998647,磐田市香りの博物館,museum,34.7154870,137.8195482
1421000025,芸術の森ろう人形美術館＆メキシコ館,museum,34.9055203,139.0883977
1421000084,菊川市黒田家代官屋敷資料館,museum,34.6986360,138.0844290
1421000378,蓮華寺文化財記念館,museum,34.8393514,137.9253320
1421000509,袋井市立浅羽郷土資料館,museum,34.7062180,137.9323740
1421000764,資生堂アートハウス,museum,34.7654549,137.9991369
1421000765,資生堂企業資料館,museum,34.7659170,138.0006918
1421001638,浄感寺長八記念館,museum,34.7494611,138.7769188
1421001669,長楽寺宝物館,museum,34.6712360,138.9424580
1421002329,静岡市文化財資料館,museum,34.9826194,138.3762567
1421002370,静岡市立登呂博物館,museum,34.9550750,138.4081860
1421002375,静岡市立芹沢ケイ介美術館,museum,34.9550750,138.4081860
1421002545,静岡県立美術館,museum,34.9919220,138.4463110
1421003182,黄金崎クリスタルパーク,museum,34.8441124,138.7684162
1421003343,MOA美術館,museum,35.1092616,139.0753345
1442336991,御浜岬,attraction,34.9775722,138.7673831
1454650707,ディスカバリーパーク焼津,theme_park,34.8226455,138.3314184
1499560062,黒船展望台,viewpoint,34.6803053,138.9537604
1537813560,静岡市美術館,museum,34.9712812,138.3891688
1760980799,対島の滝展望台,viewpoint,34.8713432,139.1094337
1761798090,加山雄三ミュージアム,museum,34.7827346,138.7686712
1761798131,,viewpoint,34.8418337,138.7632859
1761798159,,viewpoint,34.8423093,138.7611379
1761798194,黄金崎,viewpoint,34.8429695,138.7612334
1918752610,,viewpoint,35.2430337,138.8085015
1918752654,富士見峠台,viewpoint,35.2398311,138.7999760
1928618323,静岡大学キャンパスミュージアム,museum,34.9663037,138.4310196
1994612081,白倉山,attraction,35.0096537,137.7722793
1994665862,菖蒲根山,attraction,35.0496217,137.7858098
1994672534,小仏山,attraction,35.0135950,137.8231035
1994674345,和山間峠,attraction,35.0340372,137.7956535
1994945177,ハサカ山,attraction,34.9909530,137.8155961
1995205122,丸山,attraction,35.0229719,137.8196509
1995279542,橿山,attraction,35.0435638,137.8132994
1996492324,戸口山,attraction,35.0638190,137.8208496
1999668735,蜆塚遺跡,attraction,34.7132920,137.7031213
2007400809,大草山,attraction,34.7700663,137.6213872
2012190530,舘山,attraction,34.7697440,137.6139084
2012405498,,viewpoint,34.9853270,139.0944271
2014428737,東海館望楼,viewpoint,34.9716884,139.0970081
2214459617,清水港船宿記念館　末廣,museum,35.0080535,138.4911338
2270165415,足柄公園,viewpoint,35.3200758,139.0117745
2286313694,,viewpoint,34.8183200,137.6101430
2290022555,,viewpoint,35.4313970,138.5955230
2290067398,富士ミルクランド,theme_park,35.3404594,138.5942772
2395211121,灯台資料展示館,museum,35.0386873,139.1725709
2409408348,びく石,attraction,34.9606154,138.2250992
2604079404,,viewpoint,35.1590770,138.6202540
2643845455,肉月美術館,museum,34.7960503,138.9341583
2748855043,ロコミュージアム,museum,34.8254662,138.1369196
2962622244,駒門風穴,attraction,35.2507713,138.9137218
3042913544,須走五合目,attraction,35.3656453,138.7783682
3189589661,二三月峠,viewpoint,34.8170410,137.6441980
3266951854,,viewpoint,34.9727509,138.8267163
3308675352,見晴台,viewpoint,35.0905596,138.8746933
3571915693,,viewpoint,35.2830647,138.9827086
3591728633,恋人の聖地,attraction,35.0960774,139.0770691
3597476845,Otome no kane,viewpoint,35.2878821,138.9708309
3600159420,起雲閣,museum,35.0927353,139.0714359
3624179711,鷲沢風穴,attraction,34.8491933,137.7122532
3626483089,竜ヶ岩洞,attraction,34.8451678,137.6490392
3703211251,白糸の滝,attraction,35.3129910,138.5872234
3743854027,ヤマハ発動機 コミュニケーションプラザ,museum,34.7229862,137.8767449
3812047087,大楠,attraction,35.1006878,139.0676277
3836733457,滝沢展望台,viewpoint,34.8622180,137.7142270
3839096961,薩埵峠由比側駐車場,viewpoint,35.0716788,138.5412244
3921425757,,viewpoint,35.1324660,138.5109350
3924658357,豊田佐吉記念館,museum,34.7104420,137.5280050
3995008136,城ヶ崎海岸,attraction,34.8904234,139.1385810
4052257690,三岳山,viewpoint,34.8528668,137.6909833
4084707682,不知沼の池,attraction,34.8457425,139.0140981
4120979076,,viewpoint,34.7510815,138.0007138
4142642843,,viewpoint,35.0388888,138.9631503
4144262378,掛川城天守閣,attraction,34.7751871,138.0139371
4169801380,き☆らら展望台,viewpoint,35.0642971,138.8903240
4171787493,Mt. Fuji viewpoint,viewpoint,35.3276651,139.0044052
4171787602,Mt. Fuji viewpoint,viewpoint,35.3337707,139.0010190
4185413058,スクエアクライミングセンター,attraction,34.7344759,137.7579079
4202340647,恋人岬,viewpoint,34.8675494,138.7613952
4238847193,Lure's,attraction,34.7422119,137.7245323
4240778846,煌めきの丘,viewpoint,34.9965282,138.7790311
4240778853,,viewpoint,34.9822127,138.7729813
4240778855,出逢い岬,viewpoint,34.9830766,138.7701621
4240778856,尾ヶ崎ウイング,viewpoint,34.7153302,138.9877323
4240778857,展望広場,viewpoint,34.6940815,138.7651416
4240778858,旅人岬,viewpoint,34.9168724,138.7854660
4240778859,碧の丘,viewpoint,34.9501066,138.7718026
4265465389,富士宮口五合目,attraction,35.3368500,138.7330003
4279254411,東展望台,viewpoint,34.9756844,138.4670981
4316824119,Shizunami beach,attraction,34.7296252,138.2302152
4330562563,ヴァンジ彫刻庭園美術館,museum,35.1713610,138.8754906
4330578466,IZU-PHOTO-MUSEUM,museum,35.1718894,138.8738894
4330587298,井上靖文学館,museum,35.1724675,138.8688854
4350478934,,viewpoint,35.2731066,138.3284008
4355179112,吹合ノ岬,viewpoint,35.0113472,138.5326933
4359474620,三保の松原 参道入口,viewpoint,34.9993603,138.5210654
4359477662,三保の松原,viewpoint,34.9947415,138.5238018
4375036765,河津桜,viewpoint,34.7470544,138.9924272
4398799892,ドッグラン グリーンリーフ,attraction,34.7669021,137.5014031
4398868990,おちばの里親水公園,viewpoint,34.7757357,137.4918390
4404233220,伊豆ぐらんぱる公園,theme_park,34.9062751,139.1212198
4406072706,松陰像,attraction,34.6750255,138.9596527
4406074461,松陰像（まるで素浪人の様な石造）,attraction,34.6763390,138.9608507
4413144741,,viewpoint,35.0053485,138.3763354
4427541289,,viewpoint,34.9837824,138.1935898
4439305869,中瀬邸,attraction,34.7517266,138.7777218
4439305870,小原邸,attraction,34.7515461,138.7775440
4439305871,近藤邸,attraction,34.7503624,138.7777537
4441605990,あわしまマリンパーク,attraction,35.0335761,138.8900153
4443374829,静岡県埋蔵文化財センター,museum,35.1180362,138.6259239
4460480950,掛川ステンドグラス美術館,museum,34.7761348,138.0145999
4471024345,旧依田邸,museum,34.7581295,138.8297755
4515406366,,viewpoint,35.1560524,138.9914918
4522478691,,attraction,34.7459207,137.7265547
4539793241,展望テラス,viewpoint,35.3131662,138.9671529
4581049589,富士米の宮公園,viewpoint,35.1609227,138.6579645
4620650306,岩渕一里塚（西塚）,attraction,35.1482968,138.6193388
4625318186,和泉屋,attraction,35.1201751,138.6056484
4628813845,東海道由比宿おもしろ宿場館,attraction,35.1068679,138.5670391
4641769077,大諏訪松長一里塚,attraction,35.1130965,138.8265923
4641775632,沼津日枝一里塚,attraction,35.0995379,138.8685914
4644151741,笹原一里塚,attraction,35.1470814,138.9792670
4644151743,伏見一里塚,attraction,35.1105767,138.8983902
4644258835,薩埵峠展望台,viewpoint,35.0702876,138.5412636
4652538486,接待茶屋一里塚,attraction,35.1719664,139.0070612
4663300534,碧テラス,attraction,35.0120342,138.9210091
4665288973,羽鮒山展望台,viewpoint,35.2065846,138.5808923
4665288979,烏帽子山展望台,viewpoint,34.7254108,138.7401324
4668078086,草薙一里塚,attraction,35.0058784,138.4529479
4668232838,展望台,viewpoint,35.1592320,138.6117214
4676648381,怪しい少年少女博物館,museum,34.8915790,139.1169018
4680152766,松崎港を一望できる,viewpoint,34.7464599,138.7798036
4680152769,,viewpoint,35.1776663,138.6238650
4680152770,パノラマ展望台,viewpoint,35.1769236,138.6244980
4680152771,ちびっこ展望台,viewpoint,35.1764962,138.6280334
4680152772,張出しデッキ,viewpoint,35.1774977,138.6284636
4681426541,健康の森 展望台,viewpoint,34.9713503,138.7661573
4681467615,,viewpoint,35.0735495,138.4548276
4683488598,長沼一里塚,attraction,34.9893565,138.4141083
4686460792,酪農王国オラッチェ,attraction,35.1034648,139.0146991
4686731697,府中宿東見附の碑,attraction,34.9778730,138.3996431
4686919660,静岡県庁別館,viewpoint,34.9769777,138.3836752
4686919662,府中宿の碑,attraction,34.9762288,138.3922537
4688907376,府中一里塚,attraction,34.9684890,138.3737503
4692847089,中伊豆ワイナリーヒルズ,attraction,34.9836148,138.9974777
4695345419,丸子一里塚,attraction,34.9476009,138.3488823
4697413016,丸子宿京方見附,attraction,34.9501175,138.3374075
4703659161,,viewpoint,34.9687236,139.1277080
4703659162,,viewpoint,34.9691533,139.1274102
4703659163,汐吹岩,attraction,34.9691027,139.1275591
4745046921,法橋の松,viewpoint,34.7186137,137.7774735
4745046922,舟橋紀功碑,viewpoint,34.7288680,137.7993827
4745046923,天竜川の舟橋跡と木橋跡,viewpoint,34.7287085,137.8002626
4745071421,明治大帝玉座跡,viewpoint,34.7290236,137.8003761
4745071721,木船廃寺跡,viewpoint,34.7157301,137.7733414
4745071821,木船銅鐸出土地,viewpoint,34.7152274,137.7734198
4785798321,,viewpoint,35.2849576,138.9561744
4794831425,奥山氏居館跡,viewpoint,34.8398390,137.6345369
4794831722,井伊居館跡,viewpoint,34.8345275,137.6735481
4794831723,,viewpoint,34.8333876,137.6635181
4794862921,井伊家発祥の井戸,viewpoint,34.8264228,137.6684408
4795176322,天白磐座遺跡,viewpoint,34.8333959,137.6635238
4795184324,井伊家発祥の井戸,viewpoint,34.8264228,137.6684289
4796047724,西伊豆の松島,viewpoint,34.7822886,138.7656964
4802845734,空港・石雲院展望台,viewpoint,34.7953595,138.1832281
4826704674,霧香峠,viewpoint,34.9720517,138.8359071
4851295035,パチンコGranPa,attraction,34.7583030,137.7225947
4873298158,,viewpoint,35.3417693,138.5541315
4890896656,,viewpoint,35.1279635,139.0410570
4921385479,,viewpoint,34.9761982,139.0910085
4934763421,静岡カントリー 浜岡コース,viewpoint,34.6542482,138.0959239
4986006316,弁天島観光シンボルタワー,attraction,34.6867164,137.6009509
5026549585,沼田新田一里塚,attraction,35.1366708,138.7380424
5045581722,裾野カンツリー倶楽部,viewpoint,35.2248704,138.8703249
5142606732,木下惠介記念館,museum,34.7046527,137.7247957
5142606733,浜松復興記念館,museum,34.7060595,137.7262601
5156420649,石廊崎,viewpoint,34.6019509,138.8460475
5162186024,桜ヶ池,attraction,35.1833661,138.7094074
5162186123,プラザ通学路2,attraction,35.1745326,138.6999116
5162186222,ハウスミル,attraction,35.1683156,138.7025779
5162221621,プラザ通学路3,attraction,35.1722827,138.6966387
5162221921,プラザ通学路0,attraction,35.1764113,138.7060594
5170238946,ドリームプラザ観覧車,theme_park,35.0100607,138.4934948
5171186315,ちびまる子ちゃんランド,theme_park,35.0107570,138.4926514
5174478622,ゲームセンターアップル 大井川店,attraction,34.8096984,138.2733219
5174793335,ゲームセンターアップル 島田店,attraction,34.8266109,138.1873798
5226484225,,attraction,35.0970839,139.0606690
5237470113,野鳥観察スポット,attraction,34.7205633,137.6962388
5288486328,清涼峡展望台,viewpoint,35.1180262,138.1428757
5288942242,奥大井音戯の郷,museum,35.1071328,138.1361203
5302747211,,viewpoint,34.8721286,138.9225930
5302963569,小坂みかん共同農園,attraction,35.0157777,138.9319680
5304279504,あいあい岬,viewpoint,34.6096346,138.8266035
5306107811,喚声台,attraction,34.6946431,138.7550799
5318043371,見晴台,viewpoint,34.6594776,138.9840726
5319335144,恋人の聖地,attraction,35.1678292,138.1808809
5324959687,丹野平,viewpoint,34.8163578,138.8485129
5330309977,田牛ハマオモト自生地,viewpoint,34.6428964,138.9154538
5403558943,東経138度展望台,viewpoint,34.7454894,138.0004046
5406958695,海野光弘版画記念館,museum,34.8322159,138.1566811
5419526395,坂公民館,viewpoint,35.1384587,138.9622041
5420155100,新町橋,viewpoint,35.1216837,138.9260177
5432980297,,viewpoint,34.8259328,137.7431704
5450062693,,viewpoint,34.7117597,137.5287655
5450289655,展望台,viewpoint,34.9811373,137.8658220
5470500706,,viewpoint,34.6836395,137.5075283
5501216911,,viewpoint,34.9127571,138.9134506
5512860393,なかよし牧場,attraction,34.9060532,139.1008579
5512860396,カンガルーの丘,attraction,34.9063548,139.1020870
5512860401,わくわくモンキーハウス,attraction,34.9076976,139.1009493
5512860407,バッテリーカー,attraction,34.9070667,139.1003038
5512860408,チンパンジーの島,attraction,34.9063963,139.1010522
5512860413,ガチョウ広場,attraction,34.9067035,139.1007090
5512860418,リングテール島,attraction,34.9078932,139.1005692
5512861628,アニマルボートツアーズ乗り場,attraction,34.9066242,139.1013556
5512861630,バードパラダイス,attraction,34.9080997,139.1015942
5512861637,カピバラの露天風呂,attraction,34.9061793,139.1016751
5512861646,カピバラ広場,attraction,34.9081054,139.0995664
5515554268,ドルフィンファンタジー,attraction,34.9731180,139.1082994
5559201633,,viewpoint,34.9078174,137.9930585
5566208331,宝永山山頂,viewpoint,35.3435936,138.7520674
5585270825,,attraction,35.0083833,138.4974554
5588413382,山頂広場,viewpoint,34.9598391,138.2255649
5598647774,,viewpoint,34.9227167,138.3262364
5599839000,,viewpoint,34.9704365,137.9972717
5603010428,大竹公民館,attraction,35.1080340,138.9645350
5603010434,伊東市観光会館,attraction,34.9712974,139.1023954
5603023725,熱海市消防本部,attraction,35.0963143,139.0717864
5603023726,親水公園,attraction,35.0941880,139.0750710
5603023729,塩田,attraction,35.1187150,138.9891070
5603023731,殉国墓地,attraction,35.1218770,138.9718380
5603023734,網代漁港,attraction,35.0449890,139.0863020
5603023824,3分団,attraction,35.1086412,138.9651377
5603023829,網代港,attraction,35.0468150,139.0890220
5603023830,芝生公園,attraction,35.0913170,139.0788580
5603023831,与一ヘリポート,attraction,35.0170650,138.3699440
5603023832,差田グランド,attraction,34.6461700,138.8202090
5603023833,保健ｾﾝﾀｰ,attraction,34.7714410,138.7756010
5603023835,芝川スポーツ広場,attraction,35.2291180,138.5637780
5603023838,函南ゴルフ倶楽部,attraction,35.1427890,139.0027980
5603023839,木立キャンプ場,attraction,35.1488331,138.9973632
5603023844,芹田,attraction,35.1012690,138.9631980
5603023846,坂上,attraction,35.1030670,139.0043280
5603023847,加藤,attraction,35.1032310,139.0017960
5603023855,川口利宏,attraction,35.1116370,138.9777560
5603023858,奴田場公民館,attraction,35.1185140,139.0006580
5603023860,2分団,attraction,35.1041902,139.0137257
5603023861,萬城の滝キャンプ場,attraction,34.9063660,138.9942940
5603023862,東大寺,attraction,34.7597409,138.9807715
5603023863,まどが浜海遊公園,attraction,34.6758278,138.9553827
5603023864,雲見くじら館,attraction,34.7263194,138.7450991
5603023865,天窓洞,attraction,34.7827120,138.7660880
5603023867,花時計,attraction,34.9106730,138.7905860
5603023868,伊豆・三津シーパラダイス,attraction,35.0195265,138.8964791
5603023884,池彦2,attraction,35.1021316,138.8478335
5603024026,8分団,attraction,35.0450210,139.0818670
5603024028,展望回廊,viewpoint,34.7056397,137.7372293
5603024029,三本松集会所,attraction,35.1265010,138.9903890
5603024032,杉本,attraction,35.1023030,139.0101950
5603024034,江間いちご狩センター,attraction,35.0565279,138.9245446
5603024036,三島聖苑,attraction,35.1424100,138.9366720
5603024038,久料,attraction,35.0201796,138.8318412
5603024039,石神,attraction,35.1230940,138.9684230
5603024128,スポーツ公園,attraction,35.1033910,138.9747160
5603024137,静岡県消防学校,attraction,35.0702530,138.5190060
5603024138,高根ふれあい,attraction,35.3274060,138.9393580
5603024140,大型･小型,attraction,35.2282070,138.5637090
5603024141,ｸｰﾎﾟｰﾙ会館,attraction,34.9735616,138.3855186
5603024142,静岡ｶﾞｰﾃﾞﾝﾎﾃﾙ,attraction,34.9733530,138.3846640
5603024145,エアパーク,attraction,34.7475788,137.7116447
5603024146,かんなみ仏の里美術館,museum,35.1200650,138.9727840
5603024147,桑原公民館,attraction,35.1204210,138.9730330
5603024148,東海機工,attraction,35.1241430,138.9784810
5603024149,もちや,attraction,35.3712617,138.5766727
5603024153,高橋,attraction,35.0953250,139.0172940
5603024154,菊地,attraction,35.0683580,138.9172580
5603024159,不動の滝,attraction,35.1143700,138.9792040
5603024164,函南町焼却場,attraction,35.1194810,139.0009110
5603024172,萬城の滝,attraction,34.9058348,138.9931904
5603024173,旧天城トンネル,attraction,34.8302392,138.9357131
5603024174,下田海中水族館,attraction,34.6661868,138.9461816
5603024176,堂ヶ島観光船乗り場,attraction,34.7818720,138.7673660
5603024177,山の神,attraction,35.1174860,138.9762260
5603024191,池彦3,attraction,35.1030688,138.8480355
5603024192,池彦1,attraction,35.1018254,138.8476992
5603028222,原生の森公園,attraction,35.1536590,139.0113270
5603028227,富士総合火力演習場,attraction,35.2820460,138.8496710
5603036621,坂上,attraction,35.1040860,139.0105730
5603036721,森延彦,attraction,35.1147650,138.9804580
5604222321,,viewpoint,34.8956790,138.7666946
5604222422,富士山ビュースポット西浦江梨,viewpoint,35.0246880,138.7904925
5614122021,奥野ダム,attraction,34.9364298,139.0700847
5614329474,犬居城址望楼,viewpoint,34.9618587,137.8900232
5619654390,,viewpoint,35.0658118,138.2876265
5621787421,ﾊﾟｻﾃﾞｨﾅ公民館,attraction,35.1107760,138.9619990
5627677514,,viewpoint,34.9290149,137.6867996
5627712300,鹿路庭峠,attraction,34.9051041,139.0529998
5627712301,神代の湯,attraction,34.9484880,138.9897530
5635309780,須崎方面と伊豆諸島の眺望,viewpoint,34.6419734,138.9158602
5635309781,,viewpoint,34.6426145,138.9154217
5635309782,サンドスキー場を一望,viewpoint,34.6425383,138.9155450
5635309783,,viewpoint,34.6419006,138.9155601
5635309784,,viewpoint,34.6419712,138.9154378
5635309791,龍宮窟,attraction,34.6422393,138.9152567
5647423222,長谷川,attraction,35.1302870,138.9326790
5680470123,香音橋,attraction,35.1306157,138.9801538
5680477521,磐石橋,attraction,35.1339449,138.9828190
5712392422,長谷川,attraction,35.1376910,138.9254300
5840710252,智満寺の十本スギ,attraction,34.8927777,138.1587008
5865803536,ふわふわカーニバル,attraction,35.2911634,138.7839745
5865803537,水上アスレチック　カッパ大作戦,attraction,35.2907956,138.7833659
5865803538,ビーバーフィーバー,attraction,35.2904510,138.7836511
5906023485,,attraction,34.8810544,139.1304638
5962927022,たきや漁乗船場,attraction,34.7043209,137.6182592
6012712195,,viewpoint,34.7262900,137.6281461
6072119886,足湯,attraction,35.1034551,139.0785006
6080276119,,attraction,34.9360336,139.0681253
6080276121,,attraction,34.9364646,139.0691427
6080276123,大淀・小淀,attraction,34.8701162,139.1083853
6080276124,橋立つり橋,attraction,34.8708806,139.1073809
6080276691,,attraction,34.9353514,139.0648302
6081662087,,attraction,34.9691424,139.1267036
6084349387,俎岩,attraction,34.8797452,139.1322740
6084349388,,viewpoint,34.8799721,139.1315714
6084349389,にちょう,attraction,34.8778756,139.1296107
6084349391,,attraction,34.8707479,139.1132901
6084349392,,attraction,34.8955853,139.1325761
6084364586,,attraction,34.8711178,139.1122071
6084364587,,viewpoint,34.8711649,139.1114144
6084364588,,attraction,34.8896101,139.1388219
6084380387,,attraction,34.8906136,139.1387598
6084380388,,viewpoint,34.8681830,139.1025924
6084380485,,attraction,34.8695873,139.1021200
6089414507,,viewpoint,34.9748230,138.4659006
6089414518,吟望台,viewpoint,34.9740751,138.4661141
6089423787,,attraction,34.9698537,138.9261716
6091850288,,viewpoint,35.1403962,138.6962521
6091850289,富士山ドラゴンタワー,viewpoint,35.1363158,138.6943212
6101989291,,viewpoint,34.8456548,138.3310807
6104525587,,viewpoint,34.8624894,138.3191026
6166001004,きがん岩,attraction,34.7689724,137.6118386
6166001035,チャート海岸,viewpoint,34.7693299,137.6118942
6166001062,鶏冠岩,attraction,34.7683962,137.6103379
6166001063,富士見岩,viewpoint,34.7697694,137.6141647
6166001064,西行岩,attraction,34.7698689,137.6154914
6166001069,展望台,viewpoint,34.7698317,137.6136464
6171954163,大草山展望台,viewpoint,34.7699718,137.6203237
6172794871,浜名湖オルゴールミュージアム,museum,34.7700205,137.6204261
6174856557,筏場のわさび田,attraction,34.9002941,138.9660118
6175999167,稲取細野高原,attraction,34.7957002,139.0118787
6176057469,沼津港,attraction,35.0835817,138.8515199
6176057863,箱根十国峠ケーブルカー,attraction,35.1275889,139.0373720
6187442257,お湯かけ弁財天,attraction,34.8165690,139.0704864
6187615421,熱海芸妓見番歌舞練場,attraction,35.0948148,139.0718660
6187615422,熱海サンビーチ,attraction,35.0982401,139.0768609
6187998539,アタミロープウェイ,attraction,35.0886877,139.0777220
6187998540,熱海遊覧船サンレモ,attraction,35.0929842,139.0751085
6193880974,下田ロープウェイ,attraction,34.6788919,138.9461603
6237552897,竹林の小径,attraction,34.9697165,138.9260183
6239141404,SL資料館,museum,35.1080464,138.1375877
6246392177,修善寺虹の郷,attraction,34.9743345,138.9099213
6248746341,陶芸体験 八幡野窯,attraction,34.8734837,139.0979240
6248940475,伊豆高原ステンドグラス美術館,museum,34.9429678,139.1376141
6248940476,小室山観光リフト,attraction,34.9405486,139.1286388
6248950217,大室山,attraction,34.9044541,139.0955991
6248995174,大室山登山リフト,attraction,34.9077260,139.0967381
6249046729,伊豆のせんたんツアーセンター,attraction,34.6733169,138.9418951
6249262828,伊豆フライトハウス,attraction,34.7942322,139.0164609
6249262829,伊豆フルーツパーク,attraction,35.1307317,138.9451541
6249262830,一条竹の子村,attraction,34.6826038,138.8770420
6251050610,伊東マリンタウン,attraction,34.9853474,139.0943114
6251230164,収穫体験農園ふたつぼり,attraction,34.7712881,139.0342333
6253239518,堂ヶ島洞窟めぐり遊覧船,attraction,34.7829138,138.7680046
6253248358,堂ヶ島,attraction,34.7824095,138.7674181
6255403871,燈明ヶ崎,attraction,34.8010159,138.7552768
6255443385,かわづカーネーション見本園,attraction,34.7524087,138.9907404
6261280497,北川温泉観光協会（ムーンロード）,attraction,34.8319737,139.0746446
6261442375,西伊豆ガラス工房FARO,attraction,34.8432406,138.8011236
6261469823,黄金崎公園（馬ロック）,attraction,34.8424086,138.7602212
6265466342,自転車の国 サイクルスポーツセンター,theme_park,35.0093148,139.0118498
6265574133,浄蓮の滝,attraction,34.8727772,138.9238619
6267322175,陶磁・ガラス美術館 やんもの木,museum,34.8814618,139.1069639
6267322177,松川湖,attraction,34.9354816,139.0641722
6267385205,樋ノ口園,attraction,34.8191383,139.0509458
6267385206,丸鉄園,attraction,34.8212294,139.0547587
6275928171,ヒリゾ浜,attraction,34.6072583,138.8257723
6275928172,フィッシングパーク土肥,attraction,34.9052518,138.7836127
6280360899,道の駅開国下田みなと,attraction,34.6748064,138.9511655
6280360900,下田港内めぐり（サスケハナ号）,attraction,34.6747078,138.9527215
6280382026,村上康成美術館,museum,34.9140654,139.1086290
6282525889,山本食品 三島わさび工場,attraction,35.0764434,138.9260346
6282539443,伊東お菓子ぃ共和国,attraction,34.9699376,139.0976736
6282602804,入間千畳敷,attraction,34.6227024,138.7982103
6282638642,爪木崎俵磯,attraction,34.6587000,138.9848393
6282687718,恋人岬,attraction,34.8676628,138.7553720
6288799939,旧東海道松並木,attraction,34.6837751,137.6230356
6319608588,,attraction,35.0799696,139.0726577
6325103489,,attraction,35.0951179,139.0752856
6348493785,,attraction,34.9945368,138.5223940
6358327394,,attraction,34.7457401,138.9898191
6363828570,川坂屋,attraction,34.8026462,138.0761541
6367584093,赤猪石,attraction,34.6839137,137.6108911
6371840327,山頂,viewpoint,35.3595851,138.7311955
6373301453,馬王平,viewpoint,34.9015805,138.0628949
6402015844,,viewpoint,34.6989682,138.0825927
6417805774,シマウマ,attraction,34.7659059,137.6359080
6417805775,ダチョウ,attraction,34.7657600,137.6359428
6417805776,キリン,attraction,34.7660519,137.6358731
6417806889,小型サル舎,attraction,34.7660445,137.6345869
6417806890,アメリカバイソン,attraction,34.7662269,137.6353304
6417806891,ポニー,attraction,34.7662885,137.6350655
6417806892,ロバ,attraction,34.7663017,137.6356533
6417806914,イノシシ,attraction,34.7663829,137.6345737
6417806915,ポットベリー,attraction,34.7663598,137.6344503
6417806932,,viewpoint,34.7658970,137.6301557
6417806939,インドクジャク / シロクジャク,attraction,34.7661005,137.6339581
6417806940,セイラン / マクジャク,attraction,34.7661510,137.6338241
6417806959,ツキノワグマ,attraction,34.7665273,137.6335277
6417806960,ヒグマ,attraction,34.7666818,137.6334055
6417806991,フラミンゴ,attraction,34.7663956,137.6330925
6417807037,フンボルトペンギン,attraction,34.7665743,137.6327340
6417807049,アメリカビーバー,attraction,34.7664831,137.6325152
6417807075,カリフォルニアアシカ,attraction,34.7667283,137.6326337
6417807076,ホッキョクグマ,attraction,34.7666715,137.6323488
6417807134,観覧デッキ,viewpoint,34.7652850,137.6306251
6417807248,アンデスコンドル,attraction,34.7656953,137.6310321
6417807249,オオワシ,attraction,34.7656767,137.6309110
6417807250,オジロワシ,attraction,34.7655616,137.6309767
6417807251,ヒメコンドル,attraction,34.7656333,137.6310845
6417807252,ハクトウワシ,attraction,34.7655728,137.6310980
6417807284,ジェフロイクモザル,attraction,34.7651326,137.6305429
6417807285,エミュー,attraction,34.7651738,137.6301916
6417807286,ラマ,attraction,34.7652454,137.6302184
6417807290,クロカンガルー,attraction,34.7650355,137.6300783
6419066032,サルのアパート,attraction,34.7650427,137.6347641
6419066389,ニホンザル,attraction,34.7655291,137.6317174
6419066408,ハゴロモヅル,attraction,34.7657191,137.6321304
6419066409,マナヅル,attraction,34.7657191,137.6322641
6419066695,ライオン,attraction,34.7655915,137.6329144
6419066696,トラ,attraction,34.7655725,137.6325496
6419066703,タイリクオオカミ,attraction,34.7656982,137.6331648
6419066704,ユキヒョウ,attraction,34.7656587,137.6336014
6419066705,クロヒョウ,attraction,34.7656806,137.6334324
6419066743,アジアゾウ,attraction,34.7652883,137.6335081
6419066758,カナダヤマアラシ,attraction,34.7655733,137.6340906
6419066759,アフリカタテガミヤマアラシ,attraction,34.7656551,137.6341618
6419066760,カピバラ,attraction,34.7656178,137.6339876
6419066761,ミーアキャット,attraction,34.7654767,137.6342453
6419066762,コツメカワウソ,attraction,34.7655579,137.6342840
6419066767,レッサーパンダ,attraction,34.7656419,137.6343583
6419066768,ゴリラ,attraction,34.7656726,137.6347086
6419066769,オランウータン,attraction,34.7655420,137.6346931
6419066770,チンパンジー,attraction,34.7653985,137.6346135
6425806356,観覧車,attraction,34.7638396,137.6362044
6425806487,D51 86,attraction,34.7643116,137.6370119
6433990604,ふれあい広場,attraction,34.7666371,137.6330186
6447311464,ブルーベリーの郷,attraction,34.6980222,138.1160756
6450768297,白羽山展望台,viewpoint,35.0599775,138.0781279
6461983796,キリン 富士御殿場蒸溜所,viewpoint,35.3370687,138.8951346
6466546500,サルのアパート,attraction,34.7651572,137.6350249
6466546624,オシドリ,attraction,34.7664314,137.6332517
6466546625,ニホンジカ,attraction,34.7669117,137.6329905
6466546626,アヒル,attraction,34.7668071,137.6330091
6466546649,カモ,attraction,34.7664352,137.6330010
6469902491,きゃっしる美術館,museum,34.9798033,138.3816352
6483678138,恋人の聖地,attraction,34.7803011,137.6057493
6483693003,,viewpoint,34.7810752,137.6073655
6483804870,カナダガン,attraction,34.7663712,137.6332519
6545217637,石廊崎オーシャンパーク（南伊豆ジオパークビジターセンター）,attraction,34.6049834,138.8426985
6699521275,八形山頂見晴場,viewpoint,34.8445750,137.9209640
6699521278,,viewpoint,34.8543420,138.2339690
6739264386,かんのん浜ポットホール,attraction,34.8770316,139.1264127
6832617274,みはらしの丘,viewpoint,34.7481806,137.9753087
6832617276,エコパの丘,viewpoint,34.7453158,137.9756024
6834285686,,viewpoint,34.8950215,139.1362262
6904470839,旧植松家住宅,attraction,35.1846200,138.9043692
6909886402,荒原の棚田,attraction,34.8915880,138.9441034
6919894468,日本水泳の歴史資料室,museum,34.6754347,137.6705025
6968720825,一色の枕状溶岩,attraction,34.7915697,138.8066282
6968720826,亀甲岩,attraction,34.7709013,138.7635969
6968720829,三四郎島のトンボロ,attraction,34.7850415,138.7620924
6968720831,浮島海岸の岩脈群,attraction,34.7925945,138.7575959
7083893243,,viewpoint,34.7033932,137.6885377
7153806044,Fuji View,viewpoint,34.9789724,138.9121614
7164469886,,attraction,34.7475473,137.7120663
7260861476,うなぎパイファクトリー,attraction,34.7149657,137.6567057
7276123935,,viewpoint,34.6918785,137.6436269
7276126343,三嶋暦師の館,museum,35.1228703,138.9215505
7279206198,御殿場ルート 七合五勺,attraction,35.3522769,138.7422133
7340191678,駿河湾深海生物館,museum,34.9766949,138.7665926
7368568536,木下杢太郎記念館,museum,34.9737128,139.0959742
7391308153,イノベーションロード,museum,34.7180149,137.7357414
7431125286,,attraction,35.3615494,138.5907190
7451157347,亀岩,attraction,34.7811906,138.7664298
7477473233,,viewpoint,34.7146948,137.6243317
7658590181,お姫平,viewpoint,34.8736179,138.2517272
7658590288,富士見平展望台,viewpoint,34.8758490,138.2501265
7658590289,,viewpoint,34.8732393,138.2525318
7728004753,富士山展望台,viewpoint,35.4138463,138.5921426
7796643113,恋人の聖地,attraction,34.7699237,137.6203184
7799632393,立須,viewpoint,34.8501964,137.6994442
7800482886,ゴジラ岩,viewpoint,34.8096273,138.7625867
7802253329,大田子海岸　夕日展望所,viewpoint,34.8093660,138.7659041
7802253711,恋人岬　金の鐘,viewpoint,34.8680403,138.7564401
7802253902,船積み湾,viewpoint,34.7516196,138.7684962
7802253916,門脇吊り橋,viewpoint,34.8905542,139.1387408
7802253997,世界一の花時計,viewpoint,34.9106417,138.7906094
7802261086,ぱちんこ亭パーラー京　八木沢店,attraction,34.8915066,138.7726796
7802261090,ライオン岩,viewpoint,34.8470641,139.0814455
7805147450,竜ヶ石山のカレンフェルト,attraction,34.8479369,137.6555948
7858561260,,viewpoint,34.9342341,138.3453409
8043740416,タミヤ歴史館,museum,34.9690980,138.4192680
8132272112,,viewpoint,34.7781785,138.0178606
8133159508,,viewpoint,34.7121214,137.7255153
8310582518,,viewpoint,34.6976470,138.0337688
8310668656,,viewpoint,34.6980341,138.0371813
8318408133,,viewpoint,34.6979481,138.0341383
8326862853,人形の美術館 Soleil,museum,34.8779106,139.1128428
8450073033,白浜桜の里,attraction,34.6932207,138.9703366
8475734943,山崎プール跡,attraction,34.7106326,137.6195759
8476042619,展示室,museum,34.6972747,137.6337980
8570317470,龕附天正金鉱,attraction,34.9048938,138.7893185
8686096056,富士見平,viewpoint,35.4644394,138.1765479
8827247899,茶草場農法ビューポイント,viewpoint,34.8371182,138.0807033
8827247902,茶草場農法ビューポイント,viewpoint,34.8300758,138.0821853
8837599340,,viewpoint,35.2538957,137.9109032
8856199716,シルクロードミュージアム,museum,34.8496390,137.8364403
8893776716,吉田町展望台小山城,viewpoint,34.7788906,138.2470812
8893776717,吉田町郷土資料館,museum,34.7782120,138.2464965
9061784040,,viewpoint,35.2028111,138.8362457
9122250428,六合目,attraction,35.3697947,138.7616549
9122255502,御殿場口登山道新五合目,attraction,35.3356574,138.7951551
9187404673,バンダイホビーセンター,attraction,34.9893573,138.4115981
9223178678,三伏峠花畑,attraction,35.5549750,138.1456131
9233489072,伊東市観光会館別館,attraction,34.9717349,139.1017349
9235142127,雲見想い出岬,attraction,34.7284446,138.7422793
9235142128,雲見想い出岬,viewpoint,34.7283535,138.7422284
9269380869,MOA美術館,museum,35.1088290,139.0774695
9459694703,,viewpoint,34.9050737,139.0967733
9459694705,,viewpoint,34.9054847,139.0955639
9549478772,江川文庫 資料収蔵館,museum,35.0549714,138.9584067
9562109305,,viewpoint,35.1167891,138.6286216
9566126317,聖岳登山口,attraction,35.4158003,138.1939631
9566126417,畑薙大吊橋,attraction,35.3483650,138.1842070
9580287448,,viewpoint,35.1368891,138.7020602
9599207098,はりつけの松,attraction,34.8029068,139.0635461
9599207121,湯の華ぱぁ〜く,attraction,34.8161715,139.0686557
9703917265,愛の鐘,attraction,35.2373067,138.9285189
9782173095,展望台,viewpoint,34.6630788,137.7487643
9817358537,,viewpoint,34.8577397,137.8057731
9817358543,太平洋富士見平,viewpoint,34.8571121,137.7566671
9819743052,八幡富士展望台,viewpoint,34.8432647,138.1933764
9822708695,森の石松の墓,attraction,34.8551065,137.9166897
9831051877,雨やどり岩,viewpoint,34.7543146,137.4749520
9831097005,トキワマンサク群生地,attraction,34.7417227,137.4968040
9833556930,どうだん原,attraction,34.8823194,138.1522792
9838718571,中尾根パラダイス,viewpoint,34.7564554,137.4797169
9969055017,三島スカイウォーク,viewpoint,35.1518793,138.9806281
9984352861,GKW&Restaurant,attraction,35.2362314,138.9245413
9996585548,,viewpoint,35.3929760,138.5840857
10143616917,林泉寺の藤,attraction,34.9357816,139.0925289
10153795518,Mt Fuji View Otoge Pass,viewpoint,35.2810626,138.9888297
10218895876,,viewpoint,35.2648817,138.9765013
10235163637,熱海秘宝館,museum,35.0869181,139.0801444
10314151879,,viewpoint,34.7196155,137.7185097
10597763006,,attraction,35.1533678,138.7297595
10783711305,,viewpoint,35.2171018,138.9787256
10786387814,,viewpoint,34.6868410,138.9733256
10803373133,,viewpoint,35.2203574,138.9806578
10803648757,,viewpoint,34.8809215,139.1316369
10803648761,,attraction,34.8818050,139.1310334
10804497319,,viewpoint,34.6862878,138.9741876
10837027334,見晴台(小笠池),viewpoint,34.7347547,138.0113873
10866467374,ロックテラス,viewpoint,35.0318027,138.8892238
10907882888,富士見橋,viewpoint,34.7411915,138.0045681
11003223686,峰温泉大噴湯公園,attraction,34.7572797,138.9823154
11003223690,お万の井戸,attraction,34.7580492,138.9832903
11003223691,新町の大ソテツ,attraction,34.7580879,138.9833717
11010275860,河津桜原木,attraction,34.7563066,138.9891525
11035286691,西丹那駐車場,viewpoint,35.0850098,139.0265288
11035286701,玄岳駐車場,viewpoint,35.0846573,139.0304461
11173433813,,viewpoint,35.1360049,138.9178073
11173515638,Sumata gorge,attraction,35.1822076,138.1269184
11211393673,富士モータースポーツミュージアム,museum,35.3678006,138.9183201
11247829148,大講堂,attraction,34.9073139,139.1013067
11247829149,シャボテン狩り工房,attraction,34.9071468,139.0993406
11247829150,メキシコ館,attraction,34.9067860,139.0999307
11247829152,マダガスカル館,attraction,34.9063945,139.1003491
11247829153,森林性シャボテン館,attraction,34.9063109,139.1004135
11247829154,フクロウの森,attraction,34.9061679,139.1005396
11247829300,ロックガーデン,attraction,34.9084797,139.1003864
11247829301,太陽の広場,attraction,34.9076460,139.1013675
11250870439,浜名湖遊覧船　みっかび瀬戸港,viewpoint,34.7669952,137.5512453
11264696457,,viewpoint,35.1010305,139.0789117
11274423102,,viewpoint,34.9042479,139.1345800
11276657935,滝テラス,viewpoint,35.1450376,139.0742218
11278615803,,viewpoint,34.8923664,139.1397882
11278615826,,viewpoint,34.8891136,139.1362233
11278615840,,viewpoint,34.8898768,139.1387379
11278619869,魚見小屋,attraction,34.8952272,139.1360974
11279143654,,viewpoint,34.8713955,139.1100227
11279183085,,viewpoint,34.8716988,139.1142251
11296264564,342m見晴し良好,viewpoint,34.8541558,138.0302855
11343469796,ケ91,attraction,34.6977057,137.7217424
11366060619,,viewpoint,34.8157973,138.1338062
11569236421,,viewpoint,35.3470862,138.5599150
11569236442,,viewpoint,35.3429020,138.5562792
11569236443,,viewpoint,35.3415575,138.5591224
11569963838,,viewpoint,35.3119932,138.5882699
11569963845,,viewpoint,35.3125513,138.5875297
11699780046,,viewpoint,35.1704253,138.1811924
11801218786,岩渕一里塚（東塚）,attraction,35.1483456,138.6194878
11848589768,千鳥開館,attraction,35.0210516,138.8977546
11848602405,,viewpoint,35.0331800,138.9346826
11848606622,,viewpoint,35.0182121,138.8990847
11881558006,長島ダム展望台,viewpoint,35.1666979,138.1536247
11898413845,銚子ヶ淵,attraction,35.3303621,138.9904812
12105717423,,viewpoint,35.3648833,138.7332494
12121355645,,viewpoint,34.8910086,139.1390308
12300562886,こがねすと展望台,viewpoint,34.8415837,138.7611585
12303563079,薩埵峠,viewpoint,35.0659221,138.5393089
12303840201,,viewpoint,35.0874004,138.5481656
12304410860,,viewpoint,35.1284341,138.6127034
12343107950,静岡市歴史博物館,museum,34.9766353,138.3849981
12441135949,韮山反射炉ガイダンスセンター,museum,35.0395678,138.9615558
12540364169,塩屋跡,attraction,34.7485038,138.0415421
12560991211,秋葉古道の道標,attraction,34.8105487,137.9555130
12561328539,秋葉古道の道標,attraction,34.8151846,137.9515684
12567387865,船明の二本杉,attraction,34.8888873,137.8155849
12580529291,秋葉山本宮秋葉神社上社,attraction,34.9811091,137.8656319
12599162701,Mirador Lake Tanuki (muelle),viewpoint,35.3457664,138.5585977
12599162901,Fujisan Yumeno Ohashi,viewpoint,35.1517054,138.6759518
12599898734,羽ヶ庄字横尾パノラマ展望台,viewpoint,35.1146575,137.8296191
12606944032,杭瀬ヶ谷の池,attraction,34.8283507,137.9440434
12614937577,旅籠町,attraction,34.8356193,137.9297445
12617575446,三倉宿（秋葉古道）,attraction,34.9001782,137.9436718
12620042205,栂の木,attraction,34.9197109,137.9279092
12630082176,,viewpoint,35.0862340,139.0797201
12653927013,犬居宿,attraction,34.9561846,137.8921529
12698810411,,viewpoint,35.1234948,139.0452681
12727153589,,viewpoint,34.9396358,139.1311796
12740693976,Old train,attraction,34.8374684,138.9199705
12769684234,錦田一里塚,viewpoint,35.1282253,138.9408852
12787461580,,viewpoint,35.1546215,138.8062975
12852356601,富士山眺望スポット,attraction,34.8919629,139.0275853
12884444896,東光寺門かぶり松,attraction,34.7732108,138.0049403
12906349457,鏡の家,attraction,35.1231601,138.9105884
12908712713,縄状溶岩の露頭,attraction,35.1247436,138.9116337
12909124324,動物ふれあい広場,attraction,35.1238277,138.9104107
12909303337,縄状溶岩の露頭,attraction,35.1234843,138.9112961
12909303338,縄状溶岩の露頭,attraction,35.1228161,138.9127196
12911734157,ふれあい花広場,attraction,35.1239732,138.9098435
12911734158,のりもの広場,attraction,35.1240517,138.9105284
12911734159,どうぶつ広場,attraction,35.1238219,138.9100363
12911734190,万葉の森,attraction,35.1230971,138.9099744
12911734191,天神の森,attraction,35.1231368,138.9096215
12911734192,小浜の森,attraction,35.1246677,138.9117147
12911734193,緒明記念の森,attraction,35.1244741,138.9125155
12911734194,常盤の森,attraction,35.1232261,138.9112050
12911734195,さぎの森,attraction,35.1227332,138.9128749
13110344726,,viewpoint,34.6808859,138.9534420
13110344730,,viewpoint,34.6787165,138.9523137
13110344737,寝姿展望台,viewpoint,34.6796165,138.9528080
13130350697,堂山展望台,viewpoint,34.9875788,138.8519854
13135538801,馬場ヶ崎展望台,viewpoint,34.6703036,138.9483001
13135579801,お茶ヶ崎展望台,viewpoint,34.6665129,138.9470392
13135586701,志太ヶ浦展望台,viewpoint,34.6675673,138.9436981
13144203924,温室内わくわくイベント会場,attraction,34.7609839,138.0146809
13144203925,ふれあいフィッシュ,attraction,34.7606875,138.0144502
13144203926,インコのスイレンプール,attraction,34.7607789,138.0145280
13144203927,エボシドリとヘラサギの広場,attraction,34.7604550,138.0142343
13144203929,オシドリの池,attraction,34.7614785,138.0154373
13144203932,オオハシとトキの広場,attraction,34.7607040,138.0148190
13144203933,ハシビロコウの森,attraction,34.7603305,138.0141431
13144203934,世界のフクロウ展示,attraction,34.7617683,138.0152690
13144203935,ペンギンプール,attraction,34.7615540,138.0152475
13159444714,東御門・巽櫓,museum,34.9778971,138.3849384
13228143151,馬ノ背見晴台,viewpoint,35.2478830,138.7890259
13319499188,,viewpoint,35.0108853,138.9189898
13319499231,,viewpoint,35.0108995,138.9189452
13429854601,新幹線から見た富士山の眺め,viewpoint,35.1552711,138.7447583
35312379,秋葉古道,attraction,34.8986937,137.9435572
36709829,旧東海道,attraction,34.7633049,137.9638799
70763041,航空自衛隊 浜松広報館,museum,34.7472968,137.7113844
125750134,駿府夢広場,theme_park,34.9588243,138.4546421
126450135,,attraction,35.1811038,138.1122017
128955373,駿府匠宿,attraction,34.9533668,138.3333612
175074704,秋葉古道・塩の道,attraction,34.6882204,138.1795786
175076360,秋葉古道・塩の道,attraction,34.6898963,138.1756702
175401022,秋葉古道・塩の道,attraction,34.6975970,138.1266347
175401468,秋葉古道・塩の道,attraction,34.6991927,138.1394956
190399324,浜松まつり会館,museum,34.6656455,137.7422995
190835270,東海館,museum,34.9716626,139.0970530
191852806,秋葉道・塩の道,attraction,34.7411375,138.0614885
191853122,秋葉道・塩の道,attraction,34.7412389,138.0616175
191854333,秋葉道・塩の道,attraction,34.7411585,138.0619305
191855226,秋葉道・塩の道,attraction,34.7409540,138.0611838
194238855,浜松城,museum,34.7117702,137.7248886
194423251,賀茂真淵記念館,museum,34.7008148,137.7160651
197606690,浜松市犀ヶ崖資料館,museum,34.7193192,137.7186136
198295937,塩の道（暗闇坂）,attraction,34.7488289,138.0406796
198305533,塩の道（暗闇坂）,attraction,34.7498342,138.0391542
203732852,秋葉古道・塩の道,attraction,35.2516692,137.9100261
203734219,夢のかけ橋,attraction,34.9096415,137.8160454
203734921,秋葉古道（塩の道）,attraction,35.0915413,137.8444577
203737757,秋葉古道（塩の道）,attraction,35.0835394,137.8425357
203737827,秋葉古道・塩の道,attraction,35.2536426,137.9109433
208540632,秋葉道・塩の道,attraction,34.8259416,137.9474254
208540880,秋葉古道,attraction,34.8899368,137.9444173
208541382,秋葉古道,attraction,34.8916792,137.9463063
208542356,秋葉古道,attraction,34.8933942,137.9463116
242780257,気賀関所,museum,34.8061530,137.6486609
247425958,歴史民俗資料館,museum,35.1874023,138.6802535
247430497,東海道広重美術館,museum,35.1080092,138.5670971
247433912,江川邸,attraction,35.0546238,138.9592139
247437691,静岡市立登呂博物館,museum,34.9552806,138.4087824
250886504,浜名湖ガーデンパーク展望塔,viewpoint,34.7143209,137.6025754
262509063,掛川袋井線,attraction,34.7676674,137.9743334
263637516,スズキ歴史館,museum,34.6906877,137.6896145
278795794,キン肉マンミュージアムin沼津,museum,35.1013790,138.8585054
333230224,修善寺虹の郷,theme_park,34.9764327,138.9074370
355705247,ねむの木こども美術館どんぐり,museum,34.8286096,137.9975215
355705249,ねむの木こども美術館 ねむの木緑の中,museum,34.8294041,138.0030577
355706262,吉行淳之介文学館,museum,34.8256586,137.9943555
380034496,伊豆アンモナイト博物館,museum,34.9062812,139.1136189
385067167,牧之原市相良資料館,museum,34.6859089,138.1978625
386670252,三島スカイウォーク,theme_park,35.1519775,138.9820173
387342932,東海道舞坂宿 脇本陣,museum,34.6844006,137.6089665
389301744,久能山東照宮博物館,museum,34.9635604,138.4670081
404706297,大型展望水門 びゅうお,theme_park,35.0813209,138.8545487
411984091,芹沢光治良記念館,museum,35.0749178,138.8611811
419827461,資生堂アートハウス,museum,34.7653198,137.9989884
419827462,資生堂企業資料館,museum,34.7658512,138.0005039
422402003,掛川城移築蕗の門,attraction,34.7728712,138.0108917
426527601,浮世絵美術館　夢灯,museum,34.8158012,138.0963196
430014971,伊豆テディベア・ミュージアム,museum,34.8777367,139.1116108
435248238,とうもんの里,museum,34.6850199,137.9629251
448672049,下田開国博物館,museum,34.6731162,138.9399800
448680669,了仙寺宝物館,museum,34.6721477,138.9422555
464296572,,viewpoint,34.9700755,137.9971995
467977570,土屋氏地代官屋敷 北嶋仲右衛門家屋敷,attraction,34.8194643,137.9063522
478370383,山門,attraction,34.7369140,138.0705572
483025530,常葉美術館,museum,34.7559874,138.0812654
492168610,今切体験の里 海湖館,attraction,34.6820537,137.5894757
513812218,旧東海道,attraction,34.7616694,137.9596897
513812223,旧東海道,attraction,34.7716620,138.0215320
530835560,,museum,34.6688589,137.8396856
531606319,磐田市香りの博物館,museum,34.7154654,137.8196147
537390297,長島ダムふれあい館,museum,35.1653973,138.1515261
539367377,伊豆ぐらんぱる公園,attraction,34.9054578,139.1222188
539547693,ぐりんぱ,theme_park,35.2926721,138.7841154
557010997,舘山寺,attraction,34.7682429,137.6130493
558385334,ふじのくに茶の都ミュージアム,museum,34.8144016,138.1344201
560270839,茶房＆体験工房 さくら坂,attraction,34.8156410,139.0697661
561797544,おんやど白須賀,museum,34.6830133,137.5092561
562122669,伊豆シャボテン公園,theme_park,34.9071899,139.1007208
562172161,ねこの博物館,museum,34.8975715,139.1071927
562181614,,museum,34.8915633,139.1168336
562375327,池田20世紀美術館,museum,34.9217662,139.1055726
562375328,池田20世紀美術館,museum,34.9215402,139.1053671
562393312,伊豆一碧湖美術館,museum,34.9285500,139.1084897
564247126,熱海トリックアート迷宮館,museum,35.0864539,139.0781733
565243254,アカオハーブ＆ローズガーデン,attraction,35.0799126,139.0725550
566739239,伊豆高原アンティーク・ジュエリー・ミュージアム,museum,34.8771838,139.1216835
568598965,重要文化財 中村家住宅,attraction,34.6974943,137.6336933
568690091,蓮杖写真記念館,museum,34.6805793,138.9534891
574706324,象牙と石の彫刻美術館 ジュエルピア,museum,34.9152489,139.1195861
584061042,,viewpoint,34.9412167,138.0748750
585115676,ふるさと世界の昆虫館,museum,34.9512005,138.2527952
585573273,藤枝市郷土博物館,museum,34.8763298,138.2568779
589340514,からくり人形館,museum,34.9259870,137.6849516
600916401,有料エリア,theme_park,35.0210549,138.8638624
605958967,モン ミュゼ沼津（沼津市庄司美術館）,museum,35.0951735,138.8515492
609191392,浜松市博物館,museum,34.7125290,137.7035467
620773509,森のミュージアム,museum,35.2915807,138.7844702
621740405,富士山資料館,museum,35.2649506,138.8072640
640274856,まかいの牧場,theme_park,35.3378543,138.5849842
657169821,浮見堂,attraction,34.7670701,137.6158329
659277973,石部棚田,attraction,34.7194735,138.7641030
659517061,伊豆の長八美術館,museum,34.7487996,138.7765698
662893636,,viewpoint,34.7155860,137.6053007
662893639,,viewpoint,34.7134541,137.6008993
680687689,展望台,viewpoint,34.7052760,137.6351734
682020653,万葉資料館,museum,34.8013757,137.7605582
684948815,こんちゅう館,museum,34.7658597,137.6306234
685032777,油田の里公園資料館,museum,34.6980980,138.1599144
726581097,山本丘人記念館,museum,35.3541033,138.9294106
726581098,山本丘人記念館,museum,35.3540737,138.9296625
727211911,中郷館,museum,35.3284231,138.9383146
729523012,復生記念館,museum,35.2385900,138.9230771
737467618,大縄公園,attraction,35.0459798,139.0811883
743545453,静岡県富士山世界遺産センター,museum,35.2237444,138.6089377
791231935,C-46,attraction,34.7476834,137.7106122
791233438,F-104J,attraction,34.7477101,137.7116519
791233439,F-86F,attraction,34.7470705,137.7118754
925478284,土肥金山,theme_park,34.9081433,138.7932623
935429778,中田島砂丘,attraction,34.6606663,137.7421507
1013307984,合掌づくりの家,museum,34.6720636,137.6890432
1041893293,ほっとぱ～く源泉足湯,attraction,34.8164304,139.0726571
1169897607,まぼろし博覧会,theme_park,34.9223881,139.1196198
1216999680,岩テラス,viewpoint,35.1450216,139.0729461
1216999681,段テラス,viewpoint,35.1449966,139.0734696
1218424699,富士花鳥園,theme_park,35.4086979,138.5861547
1219680346,大池橋,attraction,34.7757637,137.9961199
1287401184,腕用ポンプ展示場,museum,34.7743297,137.6871860
1355077731,青田隧道,attraction,34.7580524,138.0255365
1356595134,,attraction,34.7809005,137.9846316
1357959628,峰之澤橋,viewpoint,35.0030279,137.8358055
1358296937,秋葉古道（塩の道）,attraction,35.0855996,137.8424955
1358296941,秋葉古道（塩の道）,attraction,35.0861586,137.8423619
1358296942,秋葉古道（塩の道）,attraction,35.0860275,137.8423663
1358296943,秋葉古道（塩の道）,attraction,35.0869280,137.8422194
1358296944,秋葉古道（塩の道）,attraction,35.0864193,137.8423154
1358319661,秋葉古道（塩の道）,attraction,35.0928342,137.8463278
1359209498,名栗の東海道松並木,attraction,34.7604390,137.9571260
1362287515,治郎柿原木,attraction,34.8340352,137.9292606
1362841726,戦国夢街道,attraction,34.9119146,137.9374912
1362841727,戦国夢街道,attraction,34.9129757,137.9382308
1362867764,戦国夢街道,attraction,34.9139708,137.9384811
1363024557,秋葉道・塩の道,attraction,34.9200339,137.9270082
1366632285,九里橋,attraction,34.9593588,137.8802177
1367373729,秋葉古道・塩の道,attraction,34.6971293,138.1280154
1367918122,秋葉古道・塩の道,attraction,34.6988695,138.1212054
1370503313,秋葉道・塩の道,attraction,34.7408199,138.0609625
1372535593,旧東海道,attraction,34.7716941,138.0224252
1374826588,旧東海道,attraction,34.7719882,138.0208540
1381171197,ぬくもりの森,attraction,34.7653477,137.6572221
1395630865,富士山こどもの国,theme_park,35.2545565,138.7721989
1421268035,駿府城坤櫓,museum,34.9781699,138.3809186
1421268037,東御門橋,attraction,34.9777365,138.3851568
1429931904,森のバードハウス,attraction,34.7609470,138.0159370
1429931905,屋外バードショー,attraction,34.7606563,138.0157461
1429931909,エミュー広場,attraction,34.7607018,138.0162151
1442238891,十里木高原展望台,viewpoint,35.2529539,138.7861836
1452369457,アクアリング,viewpoint,35.0118419,138.9205306
11267790,浜名湖パルパル,theme_park,34.7633677,137.6170049
15711322,秋葉古道（塩の道）,attraction,35.3535396,138.0046853
16132610,中央構造線,attraction,35.2042980,137.1567167

"""

def add_to_database_no_duplicate():
    conn = sqlite3.connect('attraction.db')
    cursor = conn.cursor()

    # テーブル作成時に name に UNIQUE 制約を付加
    # すでにテーブルがある場合は、一旦中身を維持したいのであれば
    # 下記の INSERT OR IGNORE で「名前被り」を自動で弾くようにします
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS attractions (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL UNIQUE, -- ここをUNIQUEにする
        category TEXT,
        lat REAL NOT NULL,
        lon REAL NOT NULL
    )
    ''')

    lines = raw_data.strip().split('\n')
    insert_data = []
    
    # Python側でも一応、今回投入するリスト内の重複をチェック
    seen_names = set()
    
    for line in lines:
        parts = line.split(',')
        if len(parts) < 4: continue 
        
        try:
            lon = float(parts[-1])
            lat = float(parts[-2])
            category = parts[-3].strip()
            name = ",".join(parts[1:-3]).strip()
            osm_id = int(parts[0])
            
            # 名前なし、または今回のリスト内で既に登場した名前はスキップ
            if not name or name in seen_names:
                continue
                
            insert_data.append((osm_id, name, category, lat, lon))
            seen_names.add(name)
            
        except (ValueError, IndexError):
            continue

    # 「INSERT OR IGNORE」により：
    # 1. IDが被った場合 -> 飛ばす
    # 2. 名前(name)が被った場合 -> 飛ばす
    cursor.executemany('''
    INSERT OR IGNORE INTO attractions (id, name, category, lat, lon)
    VALUES (?, ?, ?, ?, ?)
    ''', insert_data)

    conn.commit()
    conn.close()
    print(f"追加完了！ 重複を除いた {len(insert_data)} 件を処理しました。")

if __name__ == "__main__":
    add_to_database_no_duplicate()
```
